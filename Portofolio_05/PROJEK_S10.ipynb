## <a id='toc1_3_'></a>[3) Teori](#toc0_)

<div style="font-family:Trebuchet MS; font-size:25px; padding-bottom:20px; font-weight:bold">Proses Ekstraksi Emas</div>
<div style="margin-left:2%; width:95%;text-align:justify">

Bijih yang ditambang mengalami pemrosesan primer untuk mendapatkan campuran bijih atau *feed* yang lebih kasar. Hasil dari proses itu digunakan sebagai bahan baku untuk flotasi (alias proses untuk menghasilkan konsentrat yang lebih kasar). Setelah dilakukan flotasi, hasil partikel-partikelnya pun melewati proses pemurnian dua tahap.

<div style="text-align:center">
<img src="alur_proses_ekstraksi_emas.png" width="75%" allignment="center">
</div>

Berikut adalah uraian prosesnya:

1. **Flotasi** : Campuran bijih emas dimasukkan ke dalam tangki pelampung untuk mendapatkan konsentrat Au dan ekor (*tail*) yang lebih kasar (residu produk dengan tingkat konsentrasi logam berharga yang rendah). Kestabilan proses ini dipengaruhi oleh volatilitas dan keadaan fisikokimia yang tidak optimal dari pulp flotasi (campuran partikel padat dan cairan).
2. **Pemurnian** : Konsentrat Au yang lebih kasar mengalami dua tahap pemurnian. Setelah proses pemurnian, kita pun akan memiliki tingkat konsentrasi logam akhir dan ekor (*tail*) baru.



</div>

<div style="font-family:Trebuchet MS; font-size:25px; padding-bottom:20px; font-weight:bold">Terminologi Proses Ekstraksi</div>
<div style="margin-left:2%; width:95%;text-align:justify">

Berikut di bawah ini adalah sejumlah terminologi pada proses ekstraksi emas:
* *Rougher feed* — bahan mentah untuk proses flotasi.
* *Rougher additions* (atau *reagent additions*) — reagen untuk flotasi: *Xanthate*, *Sulphate*, dan *Depressant*.
    * *Xanthate* — penggerak atau aktivator flotasi.
    * *Sulphate* — natrium sulfida, khusus untuk proses ini.
    * *Depressant* — natrium silikat.
* *Rougher process* — flotasi.
* *Rougher tails* — residu produk.
* *Float banks* — unit flotasi.
* *Cleaner process* — pemurnian.
* *Rougher Au* — konsentrat emas yang lebih kasar.
* *Final Au* — konsentrat emas akhir.
* *air amount* — volume udara.
* *fluid levels* — ketinggian fluida pada *float banks*.
* *feed size* — ukuran partikel umpan.
* *feed rate* — laju massa umpan.

</div>

<div style="font-family:Trebuchet MS; font-size:25px; padding-bottom:20px; font-weight:bold">Perhitungan Persentase Perolehan Emas</div>
<div style="margin-left:2%; width:95%;text-align:justify">

Pada proses ektraksi emas, parameter *"Persentase Perolehan Emas"* merupakan parameter yang sangat penting. Parameter ini menunjukkan rasio massa emas setelah melewati suatu proses terhadap massa emas sebelum melewati proses.

<div style="text-align:left">
<img src="rumus_perolehan.png" width="50%">
</div>

Dengan:
* *C* — persentase emas dalam konsentrat tepat setelah proses flotasi (untuk menemukan perolehan konsentrat yang lebih kasar)/setelah dilakukan pemurnian(untuk menemukan perolehan konsentrat akhir).
* *F* — persentase emas dalam feed sebelum proses flotasi (untuk menemukan perolehan konsentrat yang lebih kasar mupun untuk menemukan perolehan konsentrat akhir).
* *T* — persentase emas dalam ekor yang lebih kasar, tepat setelah proses flotasi (untuk menemukan perolehan konsentrat yang lebih kasar)/setelah dilakukan pemurnian (untuk menemukan perolehan konsentrat akhir).

</div>

<div style="font-family:Trebuchet MS; font-size:25px; padding-bottom:20px; font-weight:bold">Metrik Evaluasi Model</div>
<div style="margin-left:2%; width:95%;text-align:justify">

Pada proyek ini, model dievaluasi menggunakan metrik yang bernama sMAPE (*symmetric Mean Absolute Percentage Error*) atau kesalahan persentase absolut rata-rata simetris. Penggunaan kta "simetris" pada metrik ini disebabkan sMAPE memperhitungkan skala target maupun prediksi.

Berikut adalah persamaan sMAPE secara umum:

<div style="text-align:left">
<img src="rumus_umum_sMAPE.jpg" width="60%">
</div>

Di mana:
* $y_i$ — Nilai target untuk observasi dengan indeks $i$ dalam set yang digunakan untuk mengukur kualitas.
* $\hat{y}_i$ — Nilai prediksi untuk observasi dengan indeks $i$, misalnya, dalam sampel uji atau test sample.
* $N$ — Jumlah observasi pada sampel.

Pada proyek ini, terdapat 2 buah target yang akan diprediksi, yaitu:
1. Perolehan konsentrat yang lebih kasar `rougher.output.recovery`.
2. Perolehan konsentrat akhir `final.output.recovery`.

Sehingga, perhitungan sMAPE yang akan digunakan pada proyek ini adalah perhitungan sMAPE gabungan yang mencakup dua nilai di atas. Berikut adalah persamaannya:

### <a id='toc4_2_2_'></a>[**Menangani *Missing-value***](#toc0_)

Berikut di bawah ini akan dilakukan pengisian *missing value* untuk seluruh kolom terkecuali kolom data **Persentase Perolehan Emas**, menggunakan metode interpolasi berbasis waktu perolehan data untuk ketiga dataset.

```python
# mengisi missing value menggunakan interpolasi untuk full dataset
selected_cols = full_df.columns.difference(['final.output.recovery','rougher.output.recovery'])
full_df_pre = full_df.copy()
full_df_pre[selected_cols] = full_df_pre[selected_cols].set_index('date').interpolate(method='time').reset_index()
```

```python
# mengisi missing value menggunakan interpolasi untuk testing dataset
test_df_pre = test_df.copy()
test_df_pre = test_df_pre.set_index('date').interpolate(method='time').reset_index()
```

Seluruh *missing value* pada fitur telah ditangani, maka selanjutnya akan dilakukan penanganan *missing value* pada kolom target, yaitu kolom `'final.output.recovery' `dan `'rougher.output.recovery'`. Penanganan *missing value* pada kolom tersebut akan dilakukan secara manual menggunakan persamaan matematis yang terdapat pada bagian **Pendahuluan**.

Disebabkan kedua kolom target tersebut tidak terdapat pada *testing dataset*, maka proses perhitungan matematis untuk kedua kolom tersebut hanya dilakukan untuk *full dataset* dan *training dataset*.

```python
# Membuat fungsi untuk mengisi missing value pada kolom target
def fillna_target(df:pd.DataFrame):
    data = df.copy()
    target_cols = ['final.output.recovery', 'rougher.output.recovery']
    C_cols_list = ['final.output.concentrate_au', 'rougher.output.concentrate_au']
    F_col = 'rougher.input.feed_au'
    T_cols_list = ['final.output.tail_au', 'rougher.output.tail_au']
    
    for i in range(len(target_cols)):
        data_na = data[data[target_cols[i]].isna()]
        C = data_na[C_cols_list[i]]
        F = data_na[F_col]
        T = data_na[T_cols_list[i]]
        data.loc[data[target_cols[i]].isna(), target_cols[i]] = (C*(F-T)/(F*(C-T)))*100
        
    # membuang data nan yang tersisa : akibat pembagian 0/0
    data = data.dropna()
    # membuang data infinity : akibat x/0 dimana x memiliki nilai
    data = data.replace([-np.inf, np.inf], np.nan).dropna()
    
    return data.reset_index(drop=True)
```

```python
# mengisi missing value pada kolom target untuk full dataset
full_df_pre = fillna_target(full_df_pre)
# mengisi missing value pada kolom target untuk training dataset
train_df_pre = fillna_target(train_df_pre)
```

# <a id='toc5_'></a>[**E. Analisa Data Eksploratif (EDA)**](#toc0_)

Berikut adalah sejumlah *custom function* yang akan digunakan pada tahap EDA.

* ***Function* untuk menggabungkan nilai sejumlah kolom menjadi sebuah baris dan memberi label nama kolom untuk setiap baris**

```python
def JoinColumns_toRows(df:pd.DataFrame, columns:list):
    concat_cols = pd.concat([df[i] for i in columns], keys=columns, names=['column', 'old_index']).reset_index()
    concat_cols = concat_cols.drop('old_index', axis='columns')
    concat_cols.columns = ['column', 'value']
    return concat_cols
```

* ***Function* untuk untuk menampilkan boxplot dan histogram**

```python
def BoxHist_plot(df:pd.DataFrame, column:str, size:tuple,hue_col:str, width:float, title:str, xlabel:str, 
                  color_list:list=None, rot:int=None, l_ylim=None, r_ylim=None, l_xlim=None, r_xlim=None):
 
    fig, ax1 = plt.subplots(ncols=2, nrows=1, figsize=size)
    # if color_list not define (the value is None)
    if color_list is None:
        color_list = ['red', 'green','blue', 'orange', 'navy', 'grey', 'pink', 'magenta']
    # unique value of the hue_col
    hue_label = df[hue_col].unique()
    
    sns.boxplot(ax=ax1[0], data=df, x=column, y=hue_col, showmeans=True,
            palette={hue_label[i]:color_list[i] for i in range(len(hue_label))},
            meanprops={"marker":"d","markerfacecolor":"white",
                       "markeredgecolor":"black","markersize":"6"})

    sns.histplot(ax=ax1[1], data=df, x=column, hue=hue_col, kde=True, binwidth=width,
            palette={hue_label[i]:color_list[i] for i in range(len(hue_label))})
    
    # Judul Utama
    fig.suptitle(title, size=15)
    
    # Anotasi Boxplot
    for elemen,num in zip(hue_label, range(len(hue_label))):
        q1 = df[df[hue_col]==elemen][column].quantile(0.25)
        q2 = df[df[hue_col]==elemen][column].quantile(0.5)
        q3 = df[df[hue_col]==elemen][column].quantile(0.75)
        minimum = df[df[hue_col]==elemen][column].min()
        maximum = df[df[hue_col]==elemen][column].max()
        iqr = q3-q1
        
        if q3+1.5*iqr<=maximum:
            top_whis = q3+1.5*iqr
        else:
            top_whis = maximum
        
        if q1-1.5*iqr>=minimum:
            bottom_whis = q1-1.5*iqr
        else:
            bottom_whis = minimum
        y_position = num-0.5
        
        color = color_list[num]
        
        ax1[0].text(q1, y_position, round(q1,1), ha='center', va='center', c=color, weight='bold', size=12,
                path_effects=[pe.withStroke(linewidth=2, foreground='white')], rotation=rot)
        ax1[0].text(q2, y_position, round(q2,1), ha='center', va='center', c=color, weight='bold', size=12,
                path_effects=[pe.withStroke(linewidth=2, foreground='white')], rotation=rot)
        ax1[0].text(q3, y_position, round(q3,1), ha='center', va='center', c=color, weight='bold', size=12,
                path_effects=[pe.withStroke(linewidth=2, foreground='white')], rotation=rot)
        ax1[0].text(top_whis, y_position, round(top_whis,1), ha='center', va='center', c=color, weight='bold', size=12,
                path_effects=[pe.withStroke(linewidth=2, foreground='white')], rotation=rot)
        if bottom_whis>0:
            ax1[0].text(bottom_whis, y_position, round(bottom_whis,1), ha='center', va='center', c=color, weight='bold', size=12,
                    path_effects=[pe.withStroke(linewidth=2, foreground='white')], rotation=rot)
        else:
            pass
        
    ax1[0].grid(alpha=0.8, linestyle='dashed')
    ax1[1].grid(alpha=0.8, linestyle='dashed')
    
    ax1[0].set_xlabel(xlabel, size=12)
    ax1[1].set_xlabel(xlabel, size=12)
    ax1[0].set_ylabel(hue_col, size=12)
    ax1[1].set_ylabel("Count", size=12)
    
    ax1[0].set_ylim(l_ylim)
    ax1[1].set_ylim(r_ylim)
    ax1[0].set_xlim(l_xlim)
    ax1[1].set_xlim(r_xlim)
    
    plt.show()
```

Berikut di bawah ini akan dilakukan pengujian hipotesis untuk memeriksa apakah persentase perolehan emas pada tahap flotasi dan tahap akhir adalah sama atau berbeda satu sama lainnya. 

Pengujian hipotesis akan dilakukan menggunakan metode uji statistik untuk 2 sampel dependen. Hal ini dilakukan karena sampel yang mengalami flotasi merupakan sampel yang sama dengan sampel yang melewati proses ekstraksi hingga tahap akhir.

Berikut di bawah ini akan dilakukan uji hipotesis yang akan menguji hipotesis berikut:
* $H_o$ **(*null hypothesis*)** &mdash; Rata-rata persentase perolehan emas pada tahap akhir sama dengan rata-rata persentase perolehan emas setelah proses flotasi.
* $H_a$ **(*alternative hypothesis*)** &mdash; Rata-rata persentase perolehan emas pada tahap akhir berbeda dengan rata-rata persentase perolehan emas setelah proses flotasi.

```python
# melakukan uji hipotesis
alpha = 0.05 # signifikansi uji hipotesis
hypo_test = st.ttest_rel(full_df_pre['rougher.output.recovery'], full_df_pre['final.output.recovery'])
pvalue = hypo_test.pvalue
print(f"P-value : {pvalue:.2%}")
print(f"Rata-rata persentase perolehan emas tahap flotasi : {full_df_pre['rougher.output.recovery'].mean()}")
print(f"Rata-rata persentase perolehan emas tahap akhir   : {full_df_pre['final.output.recovery'].mean()}")
print()
if pvalue<alpha:
    print(f"Hipotesis nol ditolak : Rata-rata persentase perolehan emas pada tahap akhir berbeda dengan rata-rata persentase perolehan emas setelah proses flotasi.")
else:
    print(f"Hipotesis nol diterima : Rata-rata persentase perolehan emas pada tahap akhir sama dengan rata-rata persentase perolehan emas setelah proses flotasi.")
    
    
```

<div style="font-size:25px; font-weight:bold; font-family:Trebuchet MS">Insights</div>
<div style="margin-left:1%; margin-top:1%; width:97%; text-align:justify">

* Rata-rata persentase perolehan emas pada tahap akhir berbeda dengan rata-rata persentase perolehan emas setelah proses flotasi.
* Persentase perolehan emas pada tahap flotasi lebih besar dari pada tahap akhir.

## <a id='toc5_4_'></a>[4) Penanganan *Outlier* dan Anomali](#toc0_)

### <a id='toc5_4_1_'></a>[**Penanganan Outlier dan Anomali Data Konsentrasi Logam**](#toc0_)

Pada distribusi data konsentrasi akumulasi logam Au-Ag-Pb ditemukan anomali dan outlier pada data. Anomali pada data berupa distribusi bimodal dimana *sub-mode* ditemukan pada range konsentrasi 0-1% sedangkan outlier ditemukan pada sisi luar *boxplot*.

Terdapatnya distribusi bimodal dapat disebabkan sejumlah alasan pada kasus ini. Distribusi bimodal tidak hanya ditemukan pada konsentrasi akumulasi logam pada setiap akhir tahap ekstraksi tetapi telah ditemukan pada kondisi awal proses (*raw-feed*). Terdapatnya distribusi bimodal pada *raw-feed* mengindikasikan penggabungan 2 populasi *raw-feed* yang berbeda sebelum proses ekstraksi dilakukan. Adapun terdapatnya distribusi bimodal pada tahap ektraksi terjadi akibat pengaturan reagen yang tidak tepat pada tahap flotasi dan pembersihan primer sehingga logam asing (non-Au-Ag-Pb) ikut tersaring bersama logam Au-Ag-Pb.

Terdapatnya distribusi bimodal pada tahap awal (*raw-feed*) merupakan suatu kesalahan parameter input yang harus disingkirkan sebelum melatih model *machine-learning*. Adapun distribusi bimodal yang terbentuk akibat pengaturan reagen yang tidak tepat merupakan suatu kejadian natural dan tidak dapat dianggap sebagai anomali. Akan tetapi pada data di kasus ini, keberadaan distribusi bimodal pada setiap tahap ektraksi juga dapat disebabkan oleh keberadaan distribusi bimodal pada tahap awal yang menjadikan rantai permasalahan pada data konsentrasi pada setiap akhir proses ekstraksi. Oleh karena itu, penanganan anomali akan dilakukan dengan cara memangkas data konsentrasi akumulasi logam pada rentang 0-1% pada kondisi awal sebelum ektraksi dimulai (*raw-feed*) dan setiap akhir proses ektraksi (flotasi, pembersihan primer dan sekunder).

Selain terdapatnya anomali, pada distribusi konsentrasi akumulasi logam juga ditemukan sejumlah outlier yang berada di sisi luar *boxplot*. Outlier yang lebih banyak pada sisi bawah dibandingkan sisi atas, mengakibatkan terbentuknya gap antara nilai median dan mean. Hadirnya outlier yang jumlahnya tidak seimbang antara sisi atas dan bawah diakibatkan faktor acak selama proses akuisisi data dilakukan sehingga outlier tersebut bukan suatu hal yang harus disingkirkan karena bersifat natural.

Berikut di bawah ini akan dilakukan proses penanganan anomali pada data konsentrasi akumulasi logam sebagaimana yang telah dijelaskan pada paragraf di atas.

* **Berikut merupakan *function* yang berfungsi untuk menangani anomali pada data konsentrasi akumulasi logam**

```python
# function untuk menangani anomali pada konsentrasi akumulasi logam
def anom_accMetal_handling(df:pd.DataFrame, thr, for_testSet=False) -> pd.DataFrame : 
    initial = ['rougher.input.feed_ag', 'rougher.input.feed_pb', 'rougher.input.feed_au']
    flot_out = ['rougher.output.concentrate_ag', 'rougher.output.concentrate_pb', 'rougher.output.concentrate_au']
    pc_out = ['primary_cleaner.output.concentrate_ag', 'primary_cleaner.output.concentrate_pb', 'primary_cleaner.output.concentrate_au']
    final_out = ['final.output.concentrate_ag', 'final.output.concentrate_pb', 'final.output.concentrate_au']
        
    if for_testSet==False:
        df = df[(df[initial].sum(axis='columns')>thr)&(df[flot_out].sum(axis='columns')>thr)&(df[pc_out].sum(axis='columns')>thr)&(df[final_out].sum(axis='columns')>thr)]
        
    else:
        df = df[df[initial].sum(axis='columns')>thr]
        
    return df
```

```python
# penanganan anomali - konsentrasi akumulasi logam
## full dataset
full_df_cln = anom_accMetal_handling(df=full_df_pre, thr=1)
## training dataset
train_df_cln = anom_accMetal_handling(df=train_df_pre, thr=1)
## testing dataset
test_df_cln = anom_accMetal_handling(df=test_df_pre, thr=1, for_testSet=True)
```

## <a id='toc6_2_'></a>[2) *Hyperparameter Tuning*](#toc0_)

Berikut merupakan sejumlah *function* yang akan digunakan pada tahap *hyperparameter tuning*.

* ***Function* untuk menghitung nilai sMAPE**

```python
# function untuk menghitung sMAPE
def sMAPE(target:pd.Series, prediction:pd.Series):
    delta_abs = abs(target-prediction)
    mean_abs = (abs(target)+abs(prediction))/2
    smape  = (delta_abs/mean_abs).mean()
    
    return smape
```

* ***Function* untuk mengevaluasi model menggunakan *cross-validation***

```python
# function untuk evaluasi model
def cv_eval(model, x, y, cv=5, show_info=True):
    kf = KFold(n_splits=cv) # menentukan pengindeksan cross-validation train/test sets
    
    mean_predict_list = [] # list yang berisi rata-rata data prediksi untuk setiap cross validasi
    mean_actual_list = [] # list yang berisi rata-rata data aktual untuk setiap cross validasi
    
    val_smape_rougher_list = [] # list yang berisi nilai sMAPE rougher untuk setiap cross validasi (prediksi validation set)
    val_smape_final_list = [] # list yang berisi nilai sMAPE final untuk setiap cross validasi (prediksi validation set)
    val_weighted_smape_list = [] # list yang berisi nilai sMAPE yg telah dibobotkan untuk setiap cross validasi (prediksi validation set)
    
    train_smape_rougher_list = [] # list yang berisi nilai sMAPE rougher untuk setiap cross validasi (prediksi training set)
    train_smape_final_list = [] # list yang berisi nilai sMAPE final untuk setiap cross validasi (prediksi training set)
    train_weighted_smape_list = [] # list yang berisi nilai sMAPE yg telah dibobotkan untuk setiap cross validasi (prediksi training set)
    
    for train_index, val_index in kf.split(y):
        # mengambil data training dan validasi
        train_y = y.iloc[train_index]
        train_x = x.iloc[train_index]
        val_y = y.iloc[val_index]
        val_x = x.iloc[val_index]
        # melatih model
        model.fit(train_x, train_y)
        # melakukan prediksi menggunakan training dan validation set
        pred_train = model.predict(train_x)
        pred_val = model.predict(val_x)
        
        # menghitung sMAPE - untuk prediksi pada validation set
        smape_final_val = sMAPE(val_y['final.output.recovery'], pred_val[:,0])
        smape_rougher_val = sMAPE(val_y['rougher.output.recovery'], pred_val[:,1])
        weighted_smape_val = 0.25*smape_rougher_val + 0.75*smape_final_val
        # menghitung sMAPE - untuk prediksi pada training set
        smape_final_train = sMAPE(train_y['final.output.recovery'], pred_train[:,0])
        smape_rougher_train = sMAPE(train_y['rougher.output.recovery'], pred_train[:,1])
        weighted_smape_train = 0.25*smape_rougher_train + 0.75*smape_final_train
        
        # memuat data
        mean_predict_list.append(np.round(pred_val.mean(axis=0),1))
        mean_actual_list.append(np.round(val_y.values.mean(axis=0),1))
        
        val_smape_rougher_list.append(smape_rougher_val)
        val_smape_final_list.append(smape_final_val)
        val_weighted_smape_list.append(weighted_smape_val)
        
        train_smape_rougher_list.append(smape_rougher_train)
        train_smape_final_list.append(smape_final_train)
        train_weighted_smape_list.append(weighted_smape_train)
    
    if show_info:
        # output
        print(f"{' Evaluasi Model ':=^70} \n")
        print(f"Avg. sMAPE - rougher  (Testing/Training) : {np.mean(val_smape_rougher_list):.2%} / {np.mean(train_smape_rougher_list):.2%}")
        print(f"Avg. sMAPE - final    (Testing/Training) : {np.mean(val_smape_final_list):.2%} / {np.mean(train_smape_final_list):.2%}")
        print(f"Avg. sMAPE - weighted (Testing/Training) : {np.mean(val_weighted_smape_list):.2%} / {np.mean(train_weighted_smape_list):.2%}")
        print(f"\n{' Prediksi ':=^70}\n")
        print(f"Avg. %perolehan emas (rougher) : {np.array(mean_predict_list)[:,1]}")
        print(f"Avg. %perolehan emas (final)   : {np.array(mean_predict_list)[:,0]}")
        print(f"\n{' Aktual ':=^70}\n")
        print(f"Avg. %perolehan emas (rougher) : {np.array(mean_actual_list)[:,1]}")
        print(f"Avg. %perolehan emas (final)   : {np.array(mean_actual_list)[:,0]}")
    else:
        pass
    
    return (np.mean(val_weighted_smape_list),np.mean(train_weighted_smape_list))
    
```

### <a id='toc6_2_2_'></a>[**Decision Tree Regressor**](#toc0_)

Pada model **Decision Tree Regressor**, masalah yang umum terjadi adalah *overfitting* akibat kedalaman pohon yang terlalu dalam sehingga menghasilkan model yang terlalu menghafal *training dataset*. Oleh karena itu perlu dilakukan *tuning* untuk mencari nilai *hyerparameter* yang tepat untuk model ini. Adapun *hyperparameter* yang akan di-*tuning* untuk model ini, yaitu `max_depth`, `min_samples_split` dan `min_samples_leaf`. 

Untuk mencari nilai *hyperparameter* yang tepat, perlu dilakukan proses iterasi dan proses evaluasi menggunakan *cross-validation*. Untuk proses *cross-validation*, dataset training akan dibagi menjadi 4 bagian (cv=4).

```python
%%time

# tuning parameter list
max_depth_list = range(2,6)
min_samples_split_list = range(2,6)
min_samples_leaf_list = range(1,6)

# best tuning parameter variable
best_dtr = None
best_max_depth = None
best_min_samples_split = None
best_min_samples_leaf = None
best_val_smape_dtr = np.inf
best_train_smape_dtr = None

# iterasi model
for max_depth in max_depth_list:
    for min_samples_split in min_samples_split_list:
        for min_samples_leaf in min_samples_leaf_list:
            if min_samples_split>min_samples_leaf:
                # model instances
                dtr = DecisionTreeRegressor(max_depth=max_depth, min_samples_split=min_samples_split,
                                            min_samples_leaf=min_samples_leaf, random_state=10)
                
                # cross validasi untuk evaluasi model
                sMape_dtr = cv_eval(dtr, train_x, train_y, cv=4, show_info=False)
                
                if sMape_dtr[0]<best_val_smape_dtr:
                    best_dtr = dtr
                    best_max_depth = max_depth
                    best_min_samples_split = min_samples_split
                    best_min_samples_leaf = min_samples_leaf
                    best_val_smape_dtr = sMape_dtr[0]
                    best_train_smape_dtr = sMape_dtr[1]
                else:
                    pass
            else:
                pass
            
# output
print(f"""
Best max_depth                         : {best_max_depth}
Best min_samples_split                 : {best_min_samples_split}
Best min_samples_leaf                  : {best_min_samples_leaf}
Best weighted sMAPE (Testing/Training) : {best_val_smape_dtr:.2%} / {best_train_smape_dtr:.2%}
      """)
            
```

### <a id='toc6_2_4_'></a>[**XGBoost Regressor**](#toc0_)

Model **XGBoost Regressor** merupakan jenis model yang menggunakan konsep *decision tree* layaknya **Random Forest Regressor** tetapi menggunakan teknik gradien boosting untuk meningkatkan performa model secara bertahap. Terdapat sejumlah *hyperparameter* yang dapat di-*tuning* agar menghasilkan model terbaik, diantaranya `n_estimators`, `learning_rate`, `max_depth`, `subsample`, dan `colsample_bytree`.

Untuk mencari nilai *hyperparameter* yang tepat, perlu dilakukan proses iterasi dan proses evaluasi menggunakan *cross-validation*. Untuk proses *cross-validation*, dataset training akan dibagi menjadi 4 bagian (cv=4).

```python
%%time

# tuning parameter list
n_estimators_list = [20, 30, 40]
learning_rate_list = [0.3, 0.1, 0.01]
max_depth_list = range(2,8)
subsample_list = [1, 0.9, 0.8]
colsample_bytree_list = [0.7, 0.6, 0.5, 0.4]

# best tuning parameter variable
best_xr = None
best_n_estimators = None
best_learning_rate = None
best_max_depth = None
best_subsample = None
best_colsample_bytree = None
best_val_smape_xr = np.inf
best_train_smape_xr = None

# iterasi model
for n_estimators in n_estimators_list:
    for learning_rate in learning_rate_list:
        for max_depth in max_depth_list:
            for subsample in subsample_list:
                for colsample_bytree in colsample_bytree_list:
                    # model instances
                    xr = XGBRegressor(n_estimators=n_estimators, learning_rate=learning_rate,
                                    max_depth=max_depth, subsample=subsample, 
                                    colsample_bytree=colsample_bytree,
                                    n_jobs=-1, random_state=10)
                    
                    # cross validasi untuk evaluasi model
                    sMape_xr = cv_eval(xr, train_x, train_y, cv=4, show_info=False)
                    
                    if sMape_xr[0]<best_val_smape_xr:
                        best_xr = xr
                        best_n_estimators = n_estimators
                        best_learning_rate = learning_rate
                        best_max_depth = max_depth
                        best_subsample = subsample
                        best_colsample_bytree = colsample_bytree
                        best_val_smape_xr = sMape_xr[0]
                        best_train_smape_xr = sMape_xr[1]
                    else:
                        pass

# output
print(f"""
Best n_estimators                      : {best_n_estimators}
Best learning_rate                     : {best_learning_rate}
Best max_depth                         : {best_max_depth}
Best subsample                         : {best_subsample}
Best colsample_bytree                  : {best_colsample_bytree}
Best weighted sMAPE (Testing/Training) : {best_val_smape_xr:.2%} / {best_train_smape_xr:.2%}
      """)
            
```

### <a id='toc6_3_1_'></a>[**Baseline Model**](#toc0_)

**Baseline Model** merupakan model yang memprediksi suatu dataset menggunakan nilai median dari *training dataset*. Model ini berguna sebagai acuan minimum kualitas suatu model yang dilatih apakah tergolong berkualitas baik atau tidak.

Berikut di bawah ini, **Baseline Model** akan melakukan prediksi untuk *testing dataset* kemudian dilakukan perhitungan skor sMAPE sebagai acuan dasar kualitas model.

```python
# prediksi training dataset
bm_pred_train = np.array(len(train_y)*[train_y.median()])
# prediksi testing dataset
bm_pred_test = np.array(len(test_y)*[train_y.median()])

# perhitungan sMAPE
train_smape_rougher = sMAPE(train_y['rougher.output.recovery'], bm_pred_train[:,1])
train_smape_final = sMAPE(train_y['final.output.recovery'], bm_pred_train[:,0])
train_weighted_smape = 0.25*train_smape_rougher + 0.75*train_smape_final

test_smape_rougher = sMAPE(test_y['rougher.output.recovery'], bm_pred_test[:,1])
test_smape_final = sMAPE(test_y['final.output.recovery'], bm_pred_test[:,0])
test_weighted_smape = 0.25*test_smape_rougher + 0.75*test_smape_final

# hasil evaluasi
print(f"{' Evaluasi Model ':=^70} \n")
print(f"Avg. sMAPE - rougher  (Testing/Training) : {test_smape_rougher:.2%} / {train_smape_rougher:.2%}")
print(f"Avg. sMAPE - final    (Testing/Training) : {test_smape_final:.2%} / {train_smape_final:.2%}")
print(f"Avg. sMAPE - weighted (Testing/Training) : {test_weighted_smape:.2%} / {train_weighted_smape:.2%}")
        
```
