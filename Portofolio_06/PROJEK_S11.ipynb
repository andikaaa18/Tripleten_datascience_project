# <a id='toc5_'></a>[**E. Analisa Data Eksploratif (EDA)**](#toc0_)

Berikut adalah sejumlah *custom function* yang akan digunakan pada tahap EDA.

* ***Function* untuk untuk menampilkan boxplot dan histogram**

```python
def BoxHist_plot(df:pd.DataFrame, column:str, size:tuple,hue_col:str, width:float, title:str, xlabel:str, kde=True,
                 discrete=False, color_list:list=None, rot:int=None, l_ylim=None, r_ylim=None, l_xlim=None, r_xlim=None):
 
    fig, ax1 = plt.subplots(ncols=2, nrows=1, figsize=size)
    # if color_list not define (the value is None)
    if color_list is None:
        color_list = ['red', 'green','blue', 'orange', 'navy', 'grey', 'pink', 'magenta']
    # unique value of the hue_col
    hue_label = df[hue_col].unique()
    
    sns.boxplot(ax=ax1[0], data=df, x=column, y=hue_col, showmeans=True,
            palette={hue_label[i]:color_list[i] for i in range(len(hue_label))},
            meanprops={"marker":"d","markerfacecolor":"white",
                       "markeredgecolor":"black","markersize":"6"})

    sns.histplot(ax=ax1[1], data=df, x=column, hue=hue_col, kde=kde, binwidth=width,
            palette={hue_label[i]:color_list[i] for i in range(len(hue_label))}, discrete=discrete)
    
    # Judul Utama
    fig.suptitle(title, size=15)
    
    # Anotasi Boxplot
    for elemen,num in zip(hue_label, range(len(hue_label))):
        q1 = df[df[hue_col]==elemen][column].quantile(0.25)
        q2 = df[df[hue_col]==elemen][column].quantile(0.5)
        q3 = df[df[hue_col]==elemen][column].quantile(0.75)
        minimum = df[df[hue_col]==elemen][column].min()
        maximum = df[df[hue_col]==elemen][column].max()
        iqr = q3-q1
        
        if q3+1.5*iqr<=maximum:
            top_whis = q3+1.5*iqr
        else:
            top_whis = maximum
        
        if q1-1.5*iqr>=minimum:
            bottom_whis = q1-1.5*iqr
        else:
            bottom_whis = minimum
        y_position = num-0.5
        
        color = color_list[num]
        
        ax1[0].text(q1, y_position, round(q1,1), ha='center', va='center', c=color, weight='bold', size=12,
                path_effects=[pe.withStroke(linewidth=2, foreground='white')], rotation=rot)
        ax1[0].text(q2, y_position, round(q2,1), ha='center', va='center', c=color, weight='bold', size=12,
                path_effects=[pe.withStroke(linewidth=2, foreground='white')], rotation=rot)
        ax1[0].text(q3, y_position, round(q3,1), ha='center', va='center', c=color, weight='bold', size=12,
                path_effects=[pe.withStroke(linewidth=2, foreground='white')], rotation=rot)
        ax1[0].text(top_whis, y_position, round(top_whis,1), ha='center', va='center', c=color, weight='bold', size=12,
                path_effects=[pe.withStroke(linewidth=2, foreground='white')], rotation=rot)
        if bottom_whis>0:
            ax1[0].text(bottom_whis, y_position, round(bottom_whis,1), ha='center', va='center', c=color, weight='bold', size=12,
                    path_effects=[pe.withStroke(linewidth=2, foreground='white')], rotation=rot)
        else:
            pass
        
    ax1[0].grid(alpha=0.8, linestyle='dashed')
    ax1[1].grid(alpha=0.8, linestyle='dashed')
    
    ax1[0].set_xlabel(xlabel, size=12)
    ax1[1].set_xlabel(xlabel, size=12)
    ax1[0].set_ylabel(hue_col, size=12)
    ax1[1].set_ylabel("Count", size=12)
    
    ax1[0].set_ylim(l_ylim)
    ax1[1].set_ylim(r_ylim)
    ax1[0].set_xlim(l_xlim)
    ax1[1].set_xlim(r_xlim)
    
    plt.show()
```

## <a id='toc7_1_'></a>[1) **Tugas 1**<br>&emsp;Menemukan Klien yang Memiliki Kemiripan dengan Klien yang Lain](#toc0_)

Pada tugas ini, kita diminta untuk membuat *tools* yang dapat memberi tahu klien mana saja yang memiliki kemiripan dengan klien tertentu di dalam dataset.

Untuk membuat *tools* tersebut, kita akan menggunakan algoritma ***k-Nearest Neighbors* (kNN)** serta membangun sebuah *function* yang akan menerima input sebuah informasi klien yang akan dicari kemiripannya pada klien yang lain.

Pada tugas ini, algoritma ***k-Nearest Neighbors* (kNN)** akan diuji dengan 4 kombinasi dari 2 kasus dan dianalisa pengaruh setiap kombinasi. Berikut adalah kombinasi yang akan diuji:
* Penskalaan
    * **Full dataset** yang tidak diskalakan.
    * **Full dataset** yang diskalakan.
* Metrik Jarak
    * Menggunakan metrik jarak **Euclidean**.
    * Menggunakan metrik jarak **Manhattan**.

Berikut di bawah ini akan dibangun model ML dengan algoritma ***k-Nearest Neighbors* (kNN)** dari nol secara manual kemudian dilanjutkan dengan membangun *function* yang dapat menerima input sebuah informasi klien yang akan dicari kemiripannya pada klien yang lain.

```python
# membuat algoritma model KNN (from scratch)

class My_KNeighborsClassifier():
    def __init__(self, n_neighbors:int=5, metric:str='euclidean') -> None:
        self.n_neighbors = n_neighbors
        self.metric = metric
        
        if self.metric not in ["manhattan", "euclidean"]:
            raise IOError
    
    def fit(self, x_train, y_train) -> None:
        self.x = np.array(x_train)
        self.y = np.array(y_train)
    
    def predict(self, x_test):
        knn_distance, knn_index = self.kneighbors(x_test, self.n_neighbors)
        return np.array([st.mode(self.y[index]).mode for index in knn_index])
    
    def kneighbors(self, x_test, n_neighbors:int):
        knn_distance = []
        knn_index = []
        
        for i in range(len(x_test)):
            if self.metric == 'euclidean':
                distance = np.sqrt(np.square(np.array(x_test)[i] - self.x).sum(axis=1))
                
            elif self.metric == 'manhattan':
                distance = abs(np.array(x_test)[i] - self.x).sum(axis=1)
                
            sort_index = distance.argsort()
            
            knn_distance.append(np.sort(distance)[:n_neighbors])
            knn_index.append(sort_index[:n_neighbors])

        return np.array(knn_distance), np.array(knn_index)
```

```python
# membuat tools untuk menyelesaikan tugas 1

def get_knn(df, n, k, metric):
    
    """
    Menampilkan k nearest neighbors

    :param df: DataFrame Pandas digunakan untuk menemukan objek-objek yang mirip di dalamnya
    :param n: nomor objek yang dicari k nearest neighbors
    :param k: jumlah k nearest neighbors yang ditampilkan
    :param metric: nama metrik jarak
    """
    feature_names = ['gender', 'age', 'income', 'family_members']
    
    nbrs = My_KNeighborsClassifier(n_neighbors=k, metric=metric)
    nbrs.fit(df[feature_names], df['insurance_benefits'])
    nbrs_distances, nbrs_indices = nbrs.kneighbors(df.iloc[[n]][feature_names], k)
    
    df_res = pd.concat([
        df.iloc[nbrs_indices[0]], 
        pd.DataFrame(nbrs_distances.T, index=nbrs_indices[0], columns=['distance'])
        ], axis=1)
    
    return df_res
```

Dalam tugas klasifikasi ini, algoritma ML yang akan digunakan adalah algoritma ***k-Nearest Neighbors* (kNN)**. Dan berdasarkan tugas sebelumnya, hasil prediksi akan bekerja lebih baik pada dataset yang telah di-*scaling* sehingga kita akan menggunakan data tersebut pada tugas ini. Adapun metrik jarak yang akan digunakan adalah metrik jarak **Manhattan** sebagaimana telah disimpulkan pada tugas sebelumnya.

Untuk menguji kualitas model ***k-Nearest Neighbors* (kNN)** tersebut kita akan membuat suatu model *dummy* yang akan menjadi *baseline model* untuk model ML sebenarnya. Model *dummy* akan memprediksi dengan cara mengeluarkan nilai 1 atau 0 secara acak dengan probabilitas tertentu.

Dalam mengevaluasi model, metrik yang akan digunakan pada tugas ini antara lain:
* **Accuracy**
* **F1 score**
* **Confusion Matrix**


Di bawah ini merupakan *function* yang akan digunakan untuk mengevaluasi model.

```python
# functin untuk evaluasi model
def eval_classifier(y_true, y_pred):
    accuracy = accuracy_score(y_true, y_pred)
    f1_score_1 = f1_score(y_true, y_pred)
    cm = (confusion_matrix(y_true, y_pred, normalize='all')*100).round(2) # dikali 100 agar dalam format percent
    
    print(f'Accuracy         : {accuracy:.2%}')
    print(f'F1 score class 1 : {f1_score_1:.2%}')
    print(f"Confusion Matrix : \n\n{pd.DataFrame(data=cm,index=['act. 0', 'act. 1'],columns=['pred. 0', 'pred. 1'])}")
```

### <a id='toc7_2_1_'></a>[**Baseline Model**](#toc0_)

Model *dummy* akan menjadi *baseline model* pada tugas ini. Model ini akan mengeluarkan nilai 1 atau 0 secara acak. Oleh karena itu, kita akan menguji model *dummy* pada sejumlah nilai probabilitas yang menghasilkan prediksi bernilai 1. Kita akan menguji dengan 4 nilai probabilitas, diantaranya:
1. Probabilitas mengeluarkan nilai 1 sebesar 0%.
2. Probabilitas mengeluarkan nilai 1 sebesar 50%.
3. Probabilitas mengeluarkan nilai 1 sebesar 100%.
4. Probabilitas mengeluarkan nilai 1 sebesar probabilitas klaim asuransi yang diterima. Probabilitas klaim asuransi yang diterima dapat didefinisikan sebagai berikut.
$$
P\{\text{klaim asuransi yang diterima}\}=\frac{\text{jumlah klien yang menerima klaim asuransi}}{\text{jumlah klien secara keseluruhan}}.
$$

Berikut di bawah ini, kita akan menghitung nilai probabilitas klaim asuransi yang diterima.

```python
# menghitung probabilitas klaim asuransi diterima
p_insurance_benefits_received = y_train['insurance_benefits_received'].mean()
print(f"Probabilitas klaim asuransi yang diterima : {p_insurance_benefits_received:.2%}")
```

Berikut di bawah ini, kita akan membangun model *dummy* untuk sejumlah nilai probabilitas mengeluarkan nilai 1 yang akan dijadikan sebagai **baseline model** pada tugas ini. Dalam pembuatan model tersebut, kita akan menggunakan distribusi binomial dalam pengambilan nilai 1 atau 0 secara acak dengan nilai probabilitas (P) tertentu.

```python
# membangun model dummy
def rnd_model_predict(P, size, seed=42):
    """
    P    : probabilitas mengeluarkan angka 1
    size : jumlah observasi test set
    """
    rng = np.random.default_rng(seed=seed)
    return rng.binomial(n=1, p=P, size=size)
```

Di bawah ini, kita akan melakukan prediksi menggunakan model *dummy* dan melakukan evaluasi model.

```python
for P in [0, y_train['insurance_benefits_received'].mean(), 0.5, 1]:

    print(f'>>> Probabilitas memperoleh "1": {P:.2%}\n')
    y_pred_rnd = rnd_model_predict(P=P, size=y_val['insurance_benefits_received'].shape[0], seed=42)
        
    eval_classifier(y_val['insurance_benefits_received'], y_pred_rnd)
    
    print(f"{'':-^80}")
```

* Dari seluruh kombinasi nilai P, akurasi tertinggi yang bisa dicapai adalah sebesar 88.38% (P = 0%) sedangkan F1 score tertinggi yang bisa tercapai adalah sebesar 20.81% (P = 100%).

## <a id='toc7_3_'></a>[3) **Tugas 3**<br>&emsp;Memprediksi Jumlah Klaim yang Akan Diajukan Klien Baru](#toc0_)

Pada tugas ini, kita akan membangun model yang dapat memprediksi jumlah klaim yang akan diajukan (data `'insurance_benefit'`) oleh klien baru dalam kurun waktu 5 tahun ke depan apabila menjadi klien perusahaan asuransi. Model akan dibangun menggunakan algoritma **Regresi Linear** dan akan dievaluasi menggunakan metrik *Root Mean Squared Error* (RMSE) dan $R_2$ score.

Model regresi linear akan dibangun dari nol (*from scratch*) dan akan dilatih dan diuji menggunakan dataset asli dan dataset yang telah di-*scaling*. Penggunaan kedua dataset pada model **Regresi Linear** akan dianalisa perbedaannya.

Berikut di bawah ini akan dibangun model **Regresi Linear** dari nol serta *function* yang akan digunakan dalam evaluasi model regresi.

```python
# membangun model regresi linear from scratch
class MyLinearRegression:
    def __init__(self):
        self.weights = None
        
    def fit(self, x_train, y_train):
        X = np.concatenate((np.ones((x_train.shape[0], 1)), x_train), axis=1)
        y = y_train
        self.weights = np.linalg.inv(X.T@X)@X.T@y
        self.w = self.weights[1:]
        self.w0 = self.weights[0]
    
    def predict(self, x_test):
        predictions = x_test@self.w + self.w0
        return predictions
```

```python
# membuat function untuk evaluasi model regresi
def eval_regressor(model, y_true, y_pred):
    rmse = math.sqrt(mean_squared_error(y_true, y_pred))
    r2 = r2_score(y_true, y_pred)
    
    print(f"RMSE                        : {rmse:.2f}")
    print(f"R2                          : {r2:.2f}")
    print(f"Linear regression coef.     : {model.w}")    
    print(f"Linear regression intercept : {model.w0}")    
```

Berikut di bawah ini akan dilakukan pelatihan dan pengujian model regresi linear menggunakan dataset asli dan dataset yang telah di-*scaling*.

```python
# Model regresi linear dengan data asli
lr = MyLinearRegression()
lr.fit(x_train, y_train['insurance_benefits'])
predictions = lr.predict(x_val)

print(f"{' Data Asli ':=^100}\n")
eval_regressor(lr, y_val['insurance_benefits'], predictions)

# Model regresi linear dengan data yang telah di-skala
lr = MyLinearRegression()
lr.fit(x_train_scl, y_train['insurance_benefits'])
predictions = lr.predict(x_val_scl)

print(f"\n{' Data yang Telah diskala ':=^100}\n")
eval_regressor(lr, y_val['insurance_benefits'], predictions)
```

## <a id='toc7_4_'></a>[4) **Tugas 4**<br>&emsp;Melakukan Pengaburan Data Tanpa Merusak Model ML](#toc0_)

Pada sesi ini, dataset akan disamarkan agar menjaga data pribadi klien. Untuk menyamarkan data-data tersebut, digunakan suatu matriks invertibel yang akan dikalikan dengan matriks fitur pada dataset. Penggunaan matriks invertibel tersebut bertujuan agar data yang telah dikaburkan dapat ditransformasi kembali pada bentuk semula. Selain itu, pengaburan data (*data obfuscation*) yang dilakukan harus dipastikan tidak merusak model yang telah dibangun sebelumnya.

### <a id='toc7_4_1_'></a>[**Pembuktian *Data Obfuscation* Bisa Bekerja Pada Linear Regression Secara Analitis**](#toc0_)

Pada bagian ini, akan dilakukan penjelasan secara analitis mengapa metode pengaburan data menggunakan matriks invertibel tidak akan mempengaruhi prediksi model regresi linear.

Algoritma model regresi linear bekerja layaknya persamaan garis lurus, dimana terdapat gradien/bobot pada variabel X dan disertai konstan (*intercept*).
$$y=mx+c$$
Persamaan di atas dapat ditulis ulang dalam bentuk yang sederhana, yaitu:
$$Y=X.w_{train}$$
dimana:
* $Y$ merupakan suatu vektor yang berisi nilai prediksi model regresi linear.
* $X$ merupakan suatu matrik yang berisi nilai fitur data yang akan diprediksi.
* $w_{train}$ merupakan suatu vektor yang berisi pembotan untuk setiap fitur + bobot bias dengan MSE paling minimum, yang diperoleh dari hasil melatih model regresi linear.

Adapun solusi untuk memperoleh nilai $w_{train}$ adalah sebagai berikut:
$$w_{train}=min_w(MSE(X.w, y))$$
dimana:
* $X.w=Y_w$ merupakan nilai prediksi model pada training set.
* $y$ merupakan nilai target sebenarnya pada training set.
* $min_w()$ merupakan suatu fungsi yang mengambil nilai bobot ($w$) dengan MSE paling minimum.

Terdapat solusi analitis untuk memperoleh nilai $w_{train}$ di atas, yaitu dengan melakukan sejumlah perhitungan matriks.
$$w_{train}=(X^TX)^{-1}X^Ty$$
dimana:
* $X$ merupakan matriks yang berisi nilai fitur untuk seluruh observasi pada *training dataset*.
* $y$ merupakan matriks nilai target sebenarnya untuk seluruh observasi pada *training dataset*.

**Apa yang Akan Terjadi Pada Prediksi Regresi Linear Bila Dilakukan *Data Obfuscation* ?**

Ketika dilakukan pengaburan data, maka matriks fitur yang semula $X$ akan berubah menjadi $X'$.
$$X'=X.P$$
Dimana $P$ merupakan matriks invertibel yang dikalikan dengan matriks $X$. Penggunaan matriks invertibel sangat penting agar dapat mentranformasi data yang telah dikaburkan kembali seperti kondisi awal. Ketika dilakukan pengaburan data, maka nilai pembobotan ($w_{train}$) akan turut berubah menjadi $w_{train, P}$.

\begin{align*}
w_{train,P} &=({X'}^TX')^{-1}{X'}^Ty \\
&=({(X.P)}^T{(X.P)})^{-1}{(X.P)}^Ty \\
&=((P^TX^T){(X.P)})^{-1}{(P^TX^T)}y \\
&=((XP)^{-1}(P^TX^T)^{-1}){(P^TX^T)}y \\
&=(P^{-1}X^{-1}(X^T)^{-1}(P^T)^{-1}){(P^TX^T)}y \\
&=P^{-1}(X^TX)^{-1}((P^T)^{-1}P^T)X^Ty \\
&=P^{-1}((X^TX)^{-1}X^Ty) \\
&=P^{-1}w_{train} \\
\end{align*}

Jika $Y$ adalah hasil prediksi model sebelum dikaburkan dan $Y'$ merupakan prediksi model setelah dikaburkan, maka:
$$Y=Xw_{train}$$
dan
\begin{align*}
Y' &= X'w_{train,P} \\
&= (XP)(P^{-1}w_{train}) \\
&= X(PP^{-1})w_{train} \\
&= Xw_{train}  \\
Y' &= Y
\end{align*}



Dari hasil di atas, dapat dibuktikan secara analitis bahwa pengaburan data menggunakan metode pengalian matriks fitur dengan matriks invertibel dapat bekerja dengan baik, dan tidak merubah hasil prediksi regresi linear.

### <a id='toc7_4_2_'></a>[**Pengujian Model Linear Regression Pada Data yang Telah Dikaburkan**](#toc0_)

Berikut di bawah ini merupakan matriks invertibel yang akan digunakan sebagai kunci enkripsi dalam melakukan pengaburan data pribadi klien. Matriks invertibel tersebut akan diperoleh secara random.

```python
# membuat matriks invertibel untuk kunci enkripsi
rng = np.random.default_rng(seed=42)
P = rng.random(size=(x_train.shape[1], x_train.shape[1]))
if np.linalg.det(P)!=0:
    print(f"{' Matrik P Bersifat Invertibel ':=^80}\n")
    print(f"P = \n{P}\n")
    print(f"P-invers = \n{np.linalg.inv(P)}")
```

Berikut di bawah ini, akan dibuat *class* yang dapat melakukan *data obfuscation*.

```python
class DataObfuscation:
    def fit(self, x):
        self.shape = (x.shape[1], x.shape[1])
        rng = np.random.default_rng(seed=42)
        self.__P = rng.random(size=self.shape) # matriks P harus di-private
    
    def transform(self, x):
        return np.array(x)@self.__P
    
    def inverse_transform(self, x):
        return np.array(x)@np.linalg.inv(self.__P)
```

*Data Obfuscation Transformer* telah dibuat sehingga kita dapat melakukan pengaburan data. Pengaburan data akan dilakukan pada fitur numerik *training dataset* maupun *validation dataset*.

```python
# membuat kunci enkripsi
key_obf = DataObfuscation()
key_obf.fit(x_train)

# melakukan obfuskasi fitur training set dan validation set
x_train_obf = pd.DataFrame(data=key_obf.transform(x_train), columns=x_train.columns)
x_val_obf = pd.DataFrame(data=key_obf.transform(x_val), columns=x_val.columns)
```

Berikut di bawah ini, akan dilakukan perbandingan antara data asli, data yang telah dikaburkan dan data yang di-retransformasi ke data asli. Kita akan memeriksa apakah *data obfuscation* berjalan dengan baik atau tidak.

```python
# membandingkan data - training set
pd.concat([x_train, 
           x_train_obf,
           pd.DataFrame(data=key_obf.inverse_transform(x_train_obf), columns=x_train_obf.columns)],
          axis=1, keys=['Original', 'Obfuscated', 'Re-transform']).sample(5, random_state=12345)
```

```python
# membandingkan data - validation set
pd.concat([x_val, 
           x_val_obf,
           pd.DataFrame(data=key_obf.inverse_transform(x_val_obf), columns=x_val_obf.columns)],
          axis=1, keys=['Original', 'Obfuscated', 'Re-transform']).sample(5, random_state=12345)
```

<div style="margin-left:1%; margin-top:1%; width:97%; text-align:justify">

* Ketika data telah dikaburkan, kita tidak dapat memahami nilai sebenarnya dari data-data tersebut. Oleh karena itu, *data obfuscation* berhasil dilakukan.
* Saat data yang telah dikaburkan, ditransformasi kembali dalam bentuk semula, maka nilai-nilai yang tidak sama dengan nol dapat dikembalikan dengan baik. Akan tetapi, untuk **nilai-nilai data asli yang bernilai nol akan dikembalikan dengan angka *float* yang mendekati nol**.

</div>

<div style="font-size:25px; font-weight:bold; font-family:Trebuchet MS">Insights</div>
<div style="margin-left:1%; margin-top:1%; width:97%; text-align:justify">

* Solusi dari hal tersebut adalah dengan melakukan normalisasi setelah transformasi dilakukan secara manual.
