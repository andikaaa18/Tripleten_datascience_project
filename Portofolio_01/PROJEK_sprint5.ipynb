**Memperbaiki Penulisan Nama Kolom**

Pada bagian sebelumnya ditemukan kolom pada dataset menggunakan kapitalisasi sehingga perlu diganti dalam format *lower case* untuk mempermudah proses pengolahan data pada bagian-bagian berikutnya.

**Memperbaiki Tipe Data pada Kolom `'user_score'`**

### <a id='toc4_2_3_'></a>[**Menangani Duplikasi Data**](#toc0_)

Pada bagian sebelumnya diketahui bahwa terdapat 2 jenis duplikasi pada dataset yang kita gunakan, yaitu duplikasi game yang memiliki atribut judul, platform, serta tahun peluncuran game yang serupa dan duplikasi game yang memiliki judul "sedikit" berbeda tetapi sebenarnya sama yang merupakan pelengkap data penjualan.

```python
# menangani duplikasi game yang memiliki atribut judul, platform, serta tahun peluncuran game yang serupa
games.drop_duplicates(["name","platform","year_of_release"], inplace=True)
games.reset_index(drop=True, inplace=True)

# menampilkan informasi data
games.info()
```

Duplikasi data jenis pertama telah ditangani, berikutnya kita akan menangani duplikat data model kedua. Duplikasi ini akan ditangani dengan cara menggabungkan sejumlah data game yang sebenarnya sama yang diindentifikasi melalui variabel judul, platform, tahun peluncuran, dan genre game.

Sejumlah game yang sama tersebut akan dilakukan proses penggabungan data penjualan. Terdapat dua kemungkinan penggabungan data, yaitu 
* **Berhasil digabungkan.** Kondisi apabila nilai penjualan antar baris game yang sama tidak terdeteksi memiliki nilai berbeda.
* **Tidak berhasil digabungkan.** Kondisi apabila nilai penjualan antar baris game yang sama terdeteksi memiliki nilai berbeda.

Apabila kondisi **Tidak berhasil digabungkan** terjadi, maka seluruh baris game tersebut akan dikeluarkan dari dataset.

Berikut adalah fungsi yang akan menangani duplikasi data jenis kedua.

```python
def duplicate_handling(df:pd.DataFrame):
    data = df.copy()
    data[['na_sales','eu_sales','jp_sales','other_sales']] = data[['na_sales','eu_sales','jp_sales','other_sales']].replace(0, np.nan)
    final = pd.DataFrame()
    problem_index = list()
    handled_index = list()
    data.loc[data['name'].str.contains("sale",case=False, na=False),"name"] = data[data['name'].str.contains("sale",case=False,na=False)]["name"].str.split(" \(", expand=True)[0]
    index_dup = data.pivot_table(index=["name","platform","year_of_release","genre"],aggfunc="size").loc[lambda x:x>1].index
    for index in index_dup:
        get_rows = data[(data["name"]==index[0])&(data["platform"]==index[1])&(data["year_of_release"]==index[2])&(data["genre"]==index[3])]
        get_index = get_rows.index
        number_rows = get_rows.shape[0]
        # checking
        indicator = []
        for i in range(number_rows):
            for j in range(i+1, number_rows):
                row1 = get_rows.reset_index(drop=True).iloc[i]
                row2 = get_rows.reset_index(drop=True).iloc[j]
                for element in ['na_sales','eu_sales','jp_sales','other_sales']:
                    equal = row1[element]==row2[element]
                    row1_na = np.isnan(row1[element])
                    row2_na = np.isnan(row2[element])
                    comb = [equal, row1_na,row2_na]
                    if (comb==[False,True,False])|(comb==[False,False,True])|(comb==[True,False,False])|(comb==[True,True,True])|(comb==[False, True, True]):
                        indicator.append(True)
                    else:
                        indicator.append(False)          
        if all(indicator):
            handled_index.extend(get_index)   
        else:
            problem_index.extend(get_index)

    final = pd.concat([data.drop(index=problem_index+handled_index),data.iloc[handled_index].groupby(
        ["name","platform","year_of_release","genre"]).agg({
            'na_sales':"mean",
            'eu_sales':"mean",
            'jp_sales':"mean",
            'other_sales':"mean",
            'critic_score':"mean",
            'user_score':"mean",
            'rating':'first'}).reset_index()])
    final[['na_sales','eu_sales','jp_sales','other_sales']] = final[['na_sales','eu_sales','jp_sales','other_sales']].replace(np.nan,0)
    # print(problem_index)
    return final.sort_values('year_of_release').reset_index(drop=True)
```

### <a id='toc5_2_3_'></a>[**Distribusi Penjualan Video Game pada Platform Populer**](#toc0_)

Pada bagian sebelumnya diketahui bahwa game non portabel merupakan platform yang lebih populer dibandingkan *console* portabel. Dan platform yang berpeluang besar untuk populer di tahun 2017 yaitu **PlayStation 4** pada urutan pertama dan **Xbox One** pada urutan kedua. Pada bagian ini, kedua platform tersebut akan diulas lebih mendalam.

```python
# Query data platform populer - PS4 dan Xbox One
pop_platform = games.query("(platform=='PS4')|(platform=='XOne')")
pop_platform.sample(5)
```

Berikut di bawah ini merupakan tabel yang berisi rata-rata penjualan global sebuah game pada setiap tahun peluncurannya, jumlah game yang diluncurkan pada masing-masing tahun, serta total penjualan global dari seluruh game pada masing-masing tahun peluncuran. Data yang digunakan adalah data-data game dengan platform yang berpotensi untuk populer di tahun 2017.

```python
# membuat tabel platform populer yang dikelompokkan berdasarkan tahun peluncuran game
year_sales_pop = pop_platform.pivot_table(index=['platform', 'year_of_release'], values='global_sales', aggfunc=['mean','count','sum']).reset_index()
year_sales_pop.columns = ['platform','year_of_release','avg_sales','num_games','total_sales']
year_sales_pop
```

Berikut di bawah ini merupakan *boxplot* yang berisi persebaran nilai penjualan global game yang dirilis pada tahun tertentu, pada platform populer. Rincian dari rata-rata penjualan global per game juga disajikan pada diagram batang pada bagian setelahnya.


```python
# boxplot distribusi penjualan global game pada platform populer untuk setiap tahun peluncuran
fig,ax1 = plt.subplots(figsize=(15,6))
sns.boxplot(ax=ax1, data=pop_platform, x='year_of_release', y='global_sales', hue='platform', palette={'XOne':'#3181ff', 'PS4':'#ffa931'})
ax1.plot(ax1.get_xticks(),
         year_sales_pop.query("platform=='XOne'").sort_values('year_of_release')['avg_sales'],
         color='navy', linewidth=2, linestyle='dashed', marker='o', label="Avg. Global Sales per Game (XOne)", alpha=0.7)
ax1.plot(ax1.get_xticks(),
         year_sales_pop.query("platform=='PS4'").sort_values('year_of_release')['avg_sales'],
         color='#82270e', linewidth=2, linestyle='dashed', marker='o', label="Avg. Global Sales per Game (PS4)", alpha=0.7)

ax1.set_ylim(0,16)

plt.grid(which='major', alpha=1, linestyle='dashed')
plt.grid(which='minor', alpha=1, linestyle='dashed')

plt.title("Distribution of Game Sales by Year of Game Launch", size=15)
plt.ylabel("Global sales (USD Million)", size=12)
plt.xlabel("Year of game launch", size=12)
plt.legend(loc='upper left')
plt.show()
```

```python
# Diagram batang penjualan rata-rata setiap game-game pada tahun peluncuran tertentu untuk setiap platform populer
fig,ax1 = plt.subplots(figsize=(15,6))
sns.barplot(ax=ax1, data=year_sales_pop,x='year_of_release',y='avg_sales', hue='platform',
            palette={'XOne':'#3181ff', 'PS4':'#ffa931'})

ax1.set_title("Average Game Sales for Popular Platforms by Year of Game Launch", size=15)
ax1.set_ylabel("Avg. global sales (USD Million)", size=12)
ax1.set_xlabel("Year of game launch", size=12)
for i in ax1.patches:
    ax1.text(x=i.get_x()+i.get_width()/2, y=i.get_height()+0.05, s=np.round(i.get_height(), 2), size=12,
             weight='bold', color='#BB2525', ha='center')
plt.grid(True, alpha=0.5, linestyle='dashed')
ax1.set_ylim(0,2.5)
plt.legend(loc='upper left')
plt.show()
```

* Secara rata-rata penjualan global per game untuk setiap game yang diluncurkan pada tahun 2013-2016, menunjukkan tren yang menurun. Hal ini menunjukkan bahwa kebanyakan game baru tidak langsung diminati saat masa awal peluncurannya sehingga nilai penjualan cenderung rendah.

Oleh karena itu, secara keseluruhan rata-rata penjualan game pada tahun 2016 jauh di bawah nilai rata-rata. Nilai rata-rata penjualan game pada tahun 2016 hanya 50% dari nilai rata-rata.

## <a id='toc6_1_'></a>[1) Menguji Hipotesis Statistik](#toc0_)

**Terdapat 2 hipotesis yang akan diuji pada bagian ini, yaitu sebagai berikut**.
1. Rata-rata ulasan pengguna platform **Xbox One** dan **PC** adalah **sama**.
2. Rata-rata ulasan pengguna genre **Action** dan **Sports** adalah **berbeda**.

Pengujian hipotesis untuk kedua hipotesis di atas akan dilakukan secara **dua arah** dan **independen** dengan *significance level* **sebesar 5%**. Pengujian dilakukan secara independen karena pengguna game antar platform atau genre berkemungkinan besar bukan orang yang sama.

### <a id='toc6_1_1_'></a>[**a. Hipotesis:** Rata-rata ulasan pengguna platform **Xbox One** dan **PC** adalah sama](#toc0_)

* **$H_o$ (*null hypothesis*)** &mdash; Rata-rata ulasan pengguna platform **Xbox One** dan **PC** adalah sama.
* **$H_a$ (*alternative hypothesis*)** &mdash; Rata-rata ulasan pengguna platform **Xbox One** dan **PC** berbeda.

Dalam melakukan uji hipotesis **independen**, kita perlu **memeriksa nilai varians** kedua platform yang ingin diuji. Pengujian varians kedua platform berguna untuk melihat homogenitas data. Pengujian varians akan dilakukan dengan metode **Levene Test** disebabkan metode ini *robust* untuk data yang tidak terdistribusi normal.

```python
# Menguji varians kedua sampel uji
alpha = 0.05 # tingkat signifikansi
var_test = st.levene(games.query("platform=='XOne'")['user_score'].dropna(), games.query("platform=='PC'")['user_score'].dropna())

print(f"P-value : {var_test.pvalue:.2%}")
if var_test.pvalue<alpha:
    print("Nilai varians ulasan pengguna PC dan Xbox One tidak homogen")
else:
    print("Nilai varians ulasan pengguna PC dan Xbox One homogen")
```

**Nilai varians ulasan pengguna pada kedua platform homogen** sehingga kedua sampel berasal dari populasi yang sama. Oleh karena itu pengujian hipotesis akan menggunakan metode **Student's T-test**. Berikut di bawah ini merupakan hasil pengujian hipotesis.

```python
# Melakukan pengujian hipotesis

alpha = 0.05 # tingkat signifikansi

hypo_test = st.ttest_ind(games.query("platform=='XOne'")['user_score'], games.query("platform=='PC'")['user_score'],
                         equal_var=True, nan_policy='omit')

print(f"P-value : {hypo_test.pvalue:.2%}")
print(f"""Rata-rata sampel ulasan pengguna pada platform Xbox One : {games.query("platform=='XOne'")['user_score'].mean()}""")
print(f"""Rata-rata sampel ulasan pengguna pada platform PC : {games.query("platform=='PC'")['user_score'].mean()}""")
print()
if hypo_test.pvalue<alpha:
    print("Hipotesis nol ditolak: Rata-rata ulasan pengguna platform Xbox One dan PC berbeda.")
else:
    print("Hipotesis diterima: Rata-rata ulasan pengguna platform Xbox One dan PC adalah sama.")
```

<div style="font-size:25px; font-weight:bold; font-family:Trebuchet MS">Findings & insights</div>
<div style="margin-left:1%; width:97%; text-align:justify">

* Berdasarkan pengujian hipotesis, diperoleh P-value sebesar 0.00%
* Dengan menggunakan tingkat signifikansi ($\alpha$) sebesar 5%, maka hipotesis nol ditolak sehingga **Rata-rata ulasan pengguna platform Xbox One dan PC berbeda**.

</div>

### <a id='toc6_1_2_'></a>[**b. Hipotesis:** Rata-rata ulasan pengguna genre **Action** dan **Sports** berbeda.](#toc0_)

* **$H_o$ (*null hypothesis*)** &mdash; Rata-rata ulasan pengguna genre **Action** dan **Sports** adalah sama.
* **$H_a$ (*alternative hypothesis*)** &mdash; Rata-rata ulasan pengguna genre **Action** dan **Sports** berbeda.

Dalam melakukan uji hipotesis **independen**, kita perlu **memeriksa nilai varians** kedua genre yang ingin diuji. Pengujian varians kedua genre berguna untuk melihat homogenitas data. Pengujian varians akan dilakukan dengan metode **Levene Test** disebabkan metode ini *robust* untuk data yang tidak terdistribusi normal.

```python
# Menguji varians kedua sampel uji
alpha = 0.05 # tingkat signifikansi
var_test = st.levene(games.query("genre=='Action'")['user_score'].dropna(), games.query("genre=='Sports'")['user_score'].dropna())

print(f"P-value : {var_test.pvalue:.2%}")
if var_test.pvalue<alpha:
    print("Nilai varians ulasan pengguna pada genre Action dan Sports tidak homogen")
else:
    print("Nilai varians ulasan pengguna pada genre Action dan Sports homogen")
```

**Nilai varians ulasan pengguna pada kedua genre berbeda (tidak homogen)** sehingga dalam pengujian hipotesis akan menggunakan metode **Welch's T-test**. Berikut di bawah ini merupakan hasil pengujian hipotesis.

```python
# Melakukan pengujian hipotesis

alpha = 0.05 # tingkat signifikansi

hypo_test = st.ttest_ind(games.query("genre=='Action'")['user_score'], games.query("genre=='Sports'")['user_score'],
                         equal_var=False, nan_policy='omit')

print(f"P-value : {hypo_test.pvalue:.2%}")
print(f"""Rata-rata sampel ulasan pengguna pada genre Action : {games.query("genre=='Action'")['user_score'].mean()}""")
print(f"""Rata-rata sampel ulasan pengguna pada genre Sports : {games.query("genre=='Sports'")['user_score'].mean()}""")
print()
if hypo_test.pvalue<alpha:
    print("Hipotesis nol ditolak: Rata-rata ulasan pengguna genre Action dan Sports berbeda.")
else:
    print("Hipotesis diterima: Rata-rata ulasan pengguna genre Action dan Sports adalah sama.")
```

<div style="font-size:25px; font-weight:bold; font-family:Trebuchet MS">Findings & insights</div>
<div style="margin-left:1%; width:97%; text-align:justify">

* Berdasarkan pengujian hipotesis, diperoleh P-value sebesar 12.87%
* Dengan menggunakan tingkat signifikansi ($\alpha$) sebesar 5%, maka hipotesis nol diterima sehingga **Rata-rata ulasan pengguna genre Action dan Sports adalah sama**.
