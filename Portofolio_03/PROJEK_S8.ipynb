### <a id='toc4_1_2_'></a>[**Memeriksa *Missing-value* pada Fitur `'Tenure'`**](#toc0_)

Berikut merupakan sejumlah baris data dengan *missing-value* pada fitur `'Tenure'`

```python
# menampilkan data dengan missing value di kolom 'Tenure'
print(f"Ukuran dataset       : {data[data['Tenure'].isna()].shape}")
print(f"Proporsi data hilang : {(data['Tenure'].isna().sum()/data.shape[0]):.2%}")
data[data['Tenure'].isna()].sample(10)
```

Data yang hilang pada fitur `'Tenure'` merupakan 9% bagian dari data dan merupakan proporsi yang cukup besar. Tidak ditemukan hubungan antara nilai hilang pada fitur `'Tenure'` dengan nilai pada fitur-fitur lainnya.

Bagaimana proporsi nilai unik pada kolom tersebut?

```python
# proporsi nilai unik pada kolom 'Tenure'
data['Tenure'].value_counts(True, dropna=False).sort_index()
```

**Temuan:**
* Proporsi pengguna yang menggunakan deposito dengan tenor 1 hingga 9 tahun menunjukkan nilai yang seragam yaitu antara 8.8% hingga 9.5% pengguna. 
* Adapun pengguna yang menggunakan deposito dengan tenor 10 tahun hanyak sebanyak 4.5% pengguna.
* Terdapat 3.8% pengguna yang menggunakan **deposito dengan tenor 0 tahun**. Apa maksud dari angka nol ini? Apakah masalah pada data atau ada alasan lainnya?


Ketika nasabah menggunakan produk deposito, nasabah akan menerima *return* berupa bunga deposito. Bunga deposito yang diperoleh nasabah dapat bervariasi tergantung sejumlah alasan, dan salah satunya karena tenor yang dipilih oleh nasabah. 

**Tenor dengan jangka yang lebih panjang akan mengakibatkan nasabah menerima bunga dengan nilai lebih tinggi tetapi likuiditasnya menurun** karena tidak dapat ditarik dalam jangka waktu singkat.

Tenor deposito dapat bervariasi, dari hitungan bulan (seperti 1 bulan, 3 bulan, atau 6 bulan) atau hitungan tahun (seperti  1 tahun, 2 tahun, 3 tahun, atau lebih). Nilai 0 pada kolom `'Tenure'` dapat disebabkan oleh penggunaan deposito dengan jangka bulanan yang tidak genap setahun sehingga ditulis dalam nilai 0 tahun. Akan tetapi terdapat kemungkinan lain, yaitu merupakan kelompok nasabah yang sama sekali tidak menggunakan produk deposito.

**Apakah *missing value* pada kolom ini terkait dengan deposito dengan tenor 0 tahun ?**

Terdapat kemungkinan bahwa nasabah yang tidak menggunakan produk deposito dicatat sebagai pengguna deposito dengan tenor 0 tahun atau dicatat sebagai nilai **`NaN`**. Hal ini didasari pada 2 hal, yaitu :
1. Proporsi nasabah yang menggunakan deposito dengan tenor 1-9 tahun seragam (8.8%-9.5%). Dan hal yang wajar bila proporsi nasabah yang menggunakan deposito dengan tenor 10 tahun cukup sedikit (4.5%) karena likuiditas deposito akan semakin menurun apabila tenor semakin panjang.
2. Proporsi nasabah yang menggunakan deposito dengan tenor kurang dari 1 tahun sebanyak 3.8%. Angka ini cukup kecil, mengingat likuiditas deposito yang cukup liquid.

Oleh karena itu, *missing value* pada kolom `'Tenure'` akan diisi dengan nilai 0 sehingga, **para nasabah dengan tenor 0 tahun merupakan representasi dari pengguna yang menggunakan deposito dengan jangka bulanan atau para nasabah yang tidak memiliki deposito sama sekali**.

# <a id='toc5_'></a>[**E. Analisa Data Eksploratif**](#toc0_)

Berikut adalah sejumlah *function* yang akan digunakan pada bagian Analisa Data Eksploratif (EDA) untuk mempermudah proses analisis.

* ***Function* yang berguna untuk membuat boxplot dan histogram pada variabel numerik.**

```python
# Function untuk membuat plot gabungan (boxplot & histogram)
def box_hist_plot(df:pd.DataFrame, column:str, size:tuple,hue_col:str, width:int, title:str, xlabel:str, 
                  rot:int=None, l_ylim=None, r_ylim=None, l_xlim=None, r_xlim=None):
    
    fig, ax1 = plt.subplots(ncols=2, nrows=1, figsize=size)
    sns.boxplot(ax=ax1[0],data=df, x=column, y=df[hue_col].astype('str'), palette={'0':'#2A9D8F', '1':'#D62828'},
                showmeans=True, meanprops={"marker":"d","markerfacecolor":"white",
                                                    "markeredgecolor":"black","markersize":"6"})
    sns.histplot(ax=ax1[1], data=df, x=column, hue=df[hue_col].astype('str'), binwidth=width, 
                 palette={'0':'#2A9D8F', '1':'#D62828'}, kde=True)
    fig.suptitle(title, size=15)

    # value boxplot
    for elemen,num in zip(df[hue_col].unique(), [0,1]):
        q1 = df[df[hue_col]==elemen][column].quantile(0.25)
        q2 = df[df[hue_col]==elemen][column].quantile(0.5)
        q3 = df[df[hue_col]==elemen][column].quantile(0.75)
        minimum = df[df[hue_col]==elemen][column].min()
        maximum = df[df[hue_col]==elemen][column].max()
        iqr = q3-q1
        
        if q3+1.5*iqr<=maximum:
            top_whis = q3+1.5*iqr
        else:
            top_whis = maximum
        
        if q1-1.5*iqr>=minimum:
            bottom_whis = q1-1.5*iqr
        else:
            bottom_whis = minimum
        y_position = num-0.5

        if elemen== df[hue_col].unique()[1]:
            color = '#2A9D8F'
        elif elemen==df[hue_col].unique()[0]:
            color = '#D62828'
        else:
            color='black'

        ax1[0].text(q1, y_position, int(q1), ha='center', va='center', c=color, weight='bold', size=12,
                path_effects=[pe.withStroke(linewidth=2, foreground='white')], rotation=rot)
        ax1[0].text(q2, y_position, int(q2), ha='center', va='center', c=color, weight='bold', size=12,
                path_effects=[pe.withStroke(linewidth=2, foreground='white')], rotation=rot)
        ax1[0].text(q3, y_position, int(q3), ha='center', va='center', c=color, weight='bold', size=12,
                path_effects=[pe.withStroke(linewidth=2, foreground='white')], rotation=rot)
        ax1[0].text(top_whis, y_position, int(top_whis), ha='center', va='center', c=color, weight='bold', size=12,
                path_effects=[pe.withStroke(linewidth=2, foreground='white')], rotation=rot)
        if bottom_whis>0:
            ax1[0].text(bottom_whis, y_position, int(bottom_whis), ha='center', va='center', c=color, weight='bold', size=12,
                    path_effects=[pe.withStroke(linewidth=2, foreground='white')], rotation=rot)
        else:
            pass

    ax1[0].grid(alpha=0.8, linestyle='dashed')
    ax1[1].grid(alpha=0.8, linestyle='dashed')
    
    ax1[0].set_xlabel(xlabel, size=12)
    ax1[1].set_xlabel(xlabel, size=12)
    
    ax1[0].set_ylim(l_ylim)
    ax1[1].set_ylim(r_ylim)
    ax1[0].set_xlim(l_xlim)
    ax1[1].set_xlim(r_xlim)
    
    plt.show()
```

* ***Function* yang berguna untuk membuat barplot pada variabel kategorik.**

```python
# Function untuk membuat barplot bagi variabel kategorik
def cat_barplot(df:pd.DataFrame, y_val:str, col:str, hue_col:str, size:tuple, title:str, 
                xlabel:str, ylabel:str, rot:int=None, ylim=None):
    fig, ax1 = plt.subplots(figsize=size)
    sns.barplot(ax=ax1, data=df, x=col, y=y_val, hue=hue_col, palette={0:'#2A9D8F', 1:'#D62828'})
    
    for i in ax1.containers[0]:
        x = i.get_x()
        y = i.get_height()
        width = i.get_width()
        ax1.text(x+width/2, y, int(y), ha='center', va='bottom', c='#2A9D8F', weight='bold', size=12,
                path_effects=[pe.withStroke(linewidth=2, foreground='white')], rotation=0)
        
    for i in ax1.containers[1]:
        x = i.get_x()
        y = i.get_height()
        width = i.get_width()
        ax1.text(x+width/2, y, int(y), ha='center', va='bottom', c='#D62828', weight='bold', size=12,
                path_effects=[pe.withStroke(linewidth=2, foreground='white')], rotation=rot)
    
    ax1.set_title(title, size=15)
    ax1.grid(alpha=0.8, linestyle='dashed')

    ax1.set_xlabel(xlabel, size=12)
    ax1.set_ylabel(ylabel, size=12)

    ax1.set_ylim(ylim)

    plt.show()
```

* ***Function* yang berguna untuk melakukan uji hipotesis pada variabel numerik.**

```python
# Function untuk uji hipotesis variabel numerik
def ind_2smpl_test_num(sample_1:pd.Series, sample_2:pd.Series, alpha=0.05, Ho:str = "Hipotesis nol diterima", Ha:str = "Hipotesis nol ditolak"):
    """ 
    Uji Distribusi normal (Shapiro Wilk Test) --> downsampling dataset jika lebih dari 5000 data
        1) Jika berdistribusi normal :
            1.1) Uji variance : equal variance atau tidak?
                > jika equal: gunakan student T-test
                > jika not equal: gunakan Welch T-test
                
        2) Jika tidak berdistribusi normal :
            > Gunakan Mann-Whitney U-test
    """
    sample_1 = sample_1.sample(5000 if sample_1.shape[0]>=5000 else sample_1.shape[0], random_state=12345)
    sample_2 = sample_2.sample(5000 if sample_2.shape[0]>=5000 else sample_2.shape[0], random_state=12345)
    
    # uji distribusi normal (shapiro wilk test)
    print(f"\n{' Uji Distribusi Normal ':-^100}")
    if (shapiro(sample_1).pvalue>=alpha)&(shapiro(sample_2).pvalue>=alpha):
        print(">>> Sample 1 dan sample 2 berdistribusi normal.")
        print(f">>> P-value Sample 1 = {shapiro(sample_1).pvalue}")
        print(f">>> P-value Sample 2 = {shapiro(sample_2).pvalue}")
        print(f"\n{' Uji Variance ':-^100}")
        if levene(sample_1, sample_2).pvalue>=alpha:
            print(">>> Sample 1 dan sample 2 memiliki variance yang sama.")
            print(f"\n{' Student T-test ':-^100}")
            if ttest_ind(sample_1, sample_2, equal_var=True).pvalue>=alpha:
                print(Ho)
            else:
                print(Ha)
        else:
            print(">>> Sample 1 dan sample 2 memiliki variance yang berbeda.")
            print(f"\n{' Welchs T-test ':-^100}")
            if ttest_ind(sample_1, sample_2, equal_var=False).pvalue>=alpha:
                print(Ho)
            else:
                print(Ha)
    else:
        print(">>> Sample 1 dan sample 2 tidak berdistribusi normal.")
        print(f">>> P-value Sample 1 = {shapiro(sample_1).pvalue}")
        print(f">>> P-value Sample 2 = {shapiro(sample_2).pvalue}")
        print(f"\n{' Mann-Whitney U-test ':-^100}")
        if mannwhitneyu(sample_1, sample_2).pvalue>=alpha:
            print(Ho)
        else:
            print(Ha)
        
```

* ***Function* yang berguna untuk melakukan uji independensi pada variabel kategorik.**

```python
# function untuk uji independensi pada variabel kategorik
def ind_2smpl_test_cat(df:pd.DataFrame, column:str, index:str, value:str, alpha=0.05, Ho:str = "Hipotesis nol diterima", Ha:str = "Hipotesis nol ditolak"):
    """
    Merupakan function yang berguna untuk menguji independensi antar kolom kategorikal.
    Uji signifikansi akan menggunakan metode Chi-Square Contigency.
    
    """
    data = df.pivot_table(index=index, columns=column, values=value, aggfunc='count')
    data = data.fillna(0) # apabila ada counting bernilai nan, maka diisi dengan nol
    
    print(f"\n{' Chi-Square Contigency Test ':-^120}")
    print(f"P-value independensi = {chi2_contingency(data).pvalue}\n")
    if chi2_contingency(data).pvalue>=alpha:
        print(Ho)
    else:
        print(Ha)
    print(120*"-")
```

### <a id='toc6_2_1_'></a>[**Logistic Regression**](#toc0_)

#### <a id='toc6_2_1_1_'></a>[a. Class Weight Hyperparameter](#toc0_)

Sebelum mengatur *hyperparameter* `class_weight` pada model agar menjadi `'balanced'`, terlebih dahulu kita akan melihat bagaimana performa model bila menggunakan dataset yang *imbalance*.

```python
# membuat fungsi untuk visualisasi confusion matrix dan ROC curve
def ConfMat_x_ROC(model, x_test:pd.DataFrame, y_test:pd.Series, threshold=0.5):
    """
    Args:
        model : Model yang telah ditraining.
        x_test (pd.DataFrame): Fitur yang ingin diprediksi.
        y_test (pd.Series): Target/nilai aktual dari prediksi x_test.
        x_train (pd.DataFrame): Fitur yang digunakan untuk training model.
        y_train (pd.Series): Target/nilai aktual yang digunakan untuk training model.
    """
    prediction_prob = model.predict_proba(x_test)[:,1]
    prediction = prediction_prob>=threshold
    fpr, tpr, threshold = roc_curve(y_test, model.predict_proba(x_test)[:,1])
    
    print(f"{' Visualisasi Confusion Matrix dan ROC Curve ':=^120}")
    
    fig, ax1 = plt.subplots(ncols=2, figsize=(15,4))
    sns.heatmap(ax=ax1[0], data=confusion_matrix(y_test, prediction), annot=True, fmt=".1f", linewidths=0.8, 
                linecolor='white', square=True, cmap='Blues_r', vmin=0, vmax=2000, xticklabels=['Stay (0)','Exit (1)'], 
                yticklabels=['Stay (0)','Exit (1)'])
    sns.lineplot(ax=ax1[1], x=fpr, y=tpr, label='Model')
    sns.lineplot(ax=ax1[1], x=[0,1], y=[0,1], linestyle="--", label='Random Classifier')

    ax1[0].set_title("Confusion Matrix", size=15, pad=15)
    ax1[0].set_xlabel("Prediction", size=12)
    ax1[0].set_ylabel("Actual", size=12)

    ax1[1].set_title("ROC Curve", size=15, pad=15)
    ax1[1].set_xlabel("False Positive Rate", size=12)
    ax1[1].set_ylabel("True Positive Rate", size=12)
    ax1[1].set_xlim(0,1)
    ax1[1].set_ylim(0,1.1)

    ax1[1].grid(alpha=0.8, linestyle='dashed')

    plt.show()
    
```

```python
# membuat fungsi untuk evaluasi model
def model_evaluation(model, x_test:pd.DataFrame, y_test:pd.Series, x_train:pd.DataFrame, y_train:pd.Series, threshold=0.5):
    """
    Args:
        model : Model yang telah ditraining.
        x_test (pd.DataFrame): Fitur yang ingin diprediksi.
        y_test (pd.Series): Target/nilai aktual dari prediksi x_test.
    """
    # evaluasi model menggunakan testing dataset
    
    prediction_prob_test = model.predict_proba(x_test)[:,1]
    prediction_test = prediction_prob_test>=threshold
    accuracy_test = accuracy_score(y_test,prediction_test)
    auc_roc_test = roc_auc_score(y_test,prediction_test)
    conf_matrix_test = confusion_matrix(y_test,prediction_test)
    precision_0_test = conf_matrix_test[0,0]/(conf_matrix_test[0,0]+conf_matrix_test[1,0])
    precision_1_test = conf_matrix_test[1,1]/(conf_matrix_test[1,1]+conf_matrix_test[0,1])
    recall_0_test = conf_matrix_test[0,0]/(conf_matrix_test[0,0]+conf_matrix_test[0,1])
    recall_1_test = conf_matrix_test[1,1]/(conf_matrix_test[1,1]+conf_matrix_test[1,0])
    f1_0_test = 2*precision_0_test*recall_0_test/(precision_0_test+recall_0_test)
    f1_1_test = 2*precision_1_test*recall_1_test/(precision_1_test+recall_1_test)
    
    # evaluasi model menggunakan training dataset
    prediction_prob_train = model.predict_proba(x_train)[:,1]
    prediction_train = prediction_prob_train>=threshold
    accuracy_train = accuracy_score(y_train,prediction_train)
    auc_roc_train = roc_auc_score(y_train,prediction_train)
    conf_matrix_train = confusion_matrix(y_train,prediction_train)
    precision_0_train = conf_matrix_train[0,0]/(conf_matrix_train[0,0]+conf_matrix_train[1,0])
    precision_1_train = conf_matrix_train[1,1]/(conf_matrix_train[1,1]+conf_matrix_train[0,1])
    recall_0_train = conf_matrix_train[0,0]/(conf_matrix_train[0,0]+conf_matrix_train[0,1])
    recall_1_train = conf_matrix_train[1,1]/(conf_matrix_train[1,1]+conf_matrix_train[1,0])
    f1_0_train = 2*precision_0_train*recall_0_train/(precision_0_train+recall_0_train)
    f1_1_train = 2*precision_1_train*recall_1_train/(precision_1_train+recall_1_train)
    
    print(f"{' Evaluasi Model ':=^120}")
    print(f"Accuracy (Test/Training)             : {accuracy_test:.2%} / {accuracy_train:.2%}")
    print(f"AUC-ROC (Test/Training)              : {auc_roc_test:.2%} / {auc_roc_train:.2%}")
    print()
    print(f"Precision class 0 (Testing/Training) : {precision_0_test:.2%} / {precision_0_train:.2%}")
    print(f"Precision class 1 (Testing/Training) : {precision_1_test:.2%} / {precision_1_train:.2%}")
    print()
    print(f"Recall class 0 (Testing/Training)    : {recall_0_test:.2%} / {recall_0_train:.2%}")
    print(f"Recall class 1 (Testing/Training)    : {recall_1_test:.2%} / {recall_1_train:.2%}")
    print()
    print(f"F1 score class 0 (Testing/Training)  : {f1_0_test:.2%} / {f1_0_train:.2%}")
    print(f"F1 score class 1 (Testing/Training)  : {f1_1_test:.2%} / {f1_1_train:.2%}")
    print(120*"=")

```

```python
# instances model
lr = LogisticRegression(solver='liblinear', random_state=12345)

# training model
lr.fit(train_features, train_target)
model_evaluation(model=lr, x_test=val_features, y_test=val_target, x_train=train_features, y_train=train_target)
ConfMat_x_ROC(model=lr, x_test=val_features, y_test=val_target)
```

#### <a id='toc6_2_1_2_'></a>[b. Downsampling](#toc0_)

Pada tahap ini, kita akan mencari setup **downsampling** terbaik dengan acuan *F1 score* paling tinggi untuk *class* 1. Pada tahap ini, model akan diatur *hyperparameter* `class_weight = 'balanced'`. Output dari bagian ini adalah dapat menemukan proporsi **downsampling** yang terbaik.

```python
# membuat function untuk downsampling
def downsampling(x_train:pd.DataFrame, y_train:pd.Series, frac:float):
    features_class_0 = x_train[y_train==0]
    features_class_1 = x_train[y_train==1]
    target_class_0 = y_train[y_train==0]
    target_class_1 = y_train[y_train==1]
    
    # class yang akan di-downsampling adalah class 0
    features_class_0 = features_class_0.sample(frac=frac, random_state=12345)
    target_class_0 = target_class_0.sample(frac=frac, random_state=12345)
    
    # menggabungkan data
    ds_features = pd.concat([features_class_0, features_class_1])
    ds_target = pd.concat([target_class_0, target_class_1])
    
    # mengaduk data
    ds_features, ds_target = shuffle(ds_features, ds_target, random_state=12345)
    
    return ds_features, ds_target
```

```python
# membuat function yang mencari setup downsampling terbaik
def best_ds(model, frac_list:list, x_train, y_train, x_test, y_test):
    best_model = None
    best_frac = 0
    best_f1_class_1 = 0
    for frac in frac_list:
        x_train_ds, y_train_ds = downsampling(x_train=x_train, y_train=y_train, frac=frac)
        model.fit(x_train_ds, y_train_ds)
        predictions = model.predict(x_test)
        f1_class_1 = f1_score(y_test, predictions)
        if f1_class_1>best_f1_class_1:
            best_model = model
            best_f1_class_1 = f1_class_1
            best_frac = frac
            
    print(f"Best fraction         : {best_frac}")
    print(f"Best F1 score class 1 : {best_f1_class_1:.2%}")
    
    return best_model
```

```python
# mencari setup downsampling terbaik
lr = LogisticRegression(solver='liblinear', class_weight='balanced', random_state=12345)
lr = best_ds(model=lr, frac_list=np.arange(0.01, 0.5, 0.01), x_train=train_features, y_train=train_target, 
        x_test=val_features, y_test=val_target) 
```

Setup **downsampling** terbaik diperoleh pada fraksi 9% data *class* 0. Adapun *F1 score* terbaik yang diperoleh sebesar 53.85% dimana pada tahap sebelumnya ketika tanpa menggunakan **downsampling** kita memperoleh **F1 score** sebesar 53.03%.

**Kesimpulan :** Penggunaan **downsampling** menghasilkan 0.82% *F1 score* yang lebih baik pada **Logistic Regression**.

#### <a id='toc6_2_1_3_'></a>[c. Upsampling](#toc0_)

Pada tahap ini, kita akan mencari setup **upsampling** terbaik dengan acuan *F1 score* paling tinggi untuk *class* 1. Pada tahap ini, model akan diatur *hyperparameter* `class_weight = 'balanced'`. Output dari bagian ini adalah dapat menemukan **faktor pengali terbaik** untuk proses **upsampling**.

```python
# membuat function untuk upsampling
def upsampling(x_train:pd.DataFrame, y_train:pd.Series, multiplier:int):
    features_class_0 = x_train[y_train==0]
    features_class_1 = x_train[y_train==1]
    target_class_0 = y_train[y_train==0]
    target_class_1 = y_train[y_train==1]
    
    # class yang akan di-upsampling adalah class 1
    us_features = pd.concat([features_class_0] + [features_class_1]*multiplier)
    us_target = pd.concat([target_class_0] + [target_class_1]*multiplier)
    
    # mengaduk data
    us_features, us_target = shuffle(us_features, us_target, random_state=12345)
    
    return us_features, us_target
```

```python
# membuat function yang mencari setup upsampling terbaik
def best_us(model, multiplier_list:list[int], x_train, y_train, x_test, y_test):
    best_model = None
    best_multiplier = 0
    best_f1_class_1 = 0
    for multiplier in multiplier_list:
        x_train_us, y_train_us = upsampling(x_train=x_train, y_train=y_train, multiplier=multiplier)
        model.fit(x_train_us, y_train_us)
        predictions = model.predict(x_test)
        f1_class_1 = f1_score(y_test, predictions)
        if f1_class_1>best_f1_class_1:
            best_model = model
            best_f1_class_1 = f1_class_1
            best_multiplier = multiplier
            
    print(f"Best multiplier         : {best_multiplier}")
    print(f"Best F1 score class 1   : {best_f1_class_1:.2%}")
    
    return best_model
```

```python
# mencari setup upsampling terbaik
lr = LogisticRegression(solver='liblinear', class_weight='balanced', random_state=12345)
lr = best_us(model=lr, multiplier_list=[1,2,3,4], x_train=train_features, y_train=train_target, 
        x_test=val_features, y_test=val_target) 
```

Setup **upsampling** terbaik diperoleh pada faktor pengali sebesar 1x data *class* 1 sehingga tidak ada perubahan *F1 score* dibandingkan sebelum dilakukan **upsampling**. Hal ini menunjukkan bahwa **upsampling** tidak menaikkan performa model.

**Kesimpulan :** Penggunaan **upsampling** tidak menaikkan performa model **Logistic Regression**.

#### <a id='toc6_2_1_4_'></a>[d. Threshold Adjustment](#toc0_)

Pada tahap ini, kita akan mencari setup **threshold** terbaik yang digunakan oleh model dengan acuan *F1 score* paling tinggi untuk *class* 1. Pada tahap ini, model akan diatur *hyperparameter* `class_weight = 'balanced'`. Output dari bagian ini adalah dapat menemukan **threshold** terbaik untuk model **Logistic Regression**.

```python
# membuat function yang mencari threshold terbaik
def best_threshold(model, threshold_list:list[float], x_train, y_train, x_test, y_test):
    best_model = None
    best_threshold = 0
    best_f1_class_1 = 0
    for threshold in threshold_list:
        model.fit(x_train, y_train)
        prediction_prob = model.predict_proba(x_test)[:,1]
        predictions = prediction_prob>=threshold
        f1_class_1 = f1_score(y_test, predictions)
        if f1_class_1>best_f1_class_1:
            best_model = model
            best_f1_class_1 = f1_class_1
            best_threshold = threshold
            
    print(f"Best threshold          : {best_threshold}")
    print(f"Best F1 score class 1   : {best_f1_class_1:.2%}")
    
    return best_model
```

```python
# mencari threshold terbaik
lr = LogisticRegression(solver='liblinear', class_weight='balanced', random_state=12345)
lr = best_threshold(model=lr, threshold_list=np.arange(0.1,1,0.1), x_train=train_features, y_train=train_target, 
        x_test=val_features, y_test=val_target) 
```

Nilai **threshold** terbaik yang diperoleh yaitu 0.5 dimana ini merupakan *default threshold* pada model sehingga tidak ada perubahan *F1 score*. Hal ini menunjukkan bahwa perubahan **threshold** tidak menaikkan performa model.

**Kesimpulan :** Perubahan nilai **threshold** tidak menaikkan performa model **Logistic Regression**.

#### <a id='toc6_2_1_5_'></a>[e. Hybrid](#toc0_)

Pada tahap ini, kita akan memadu-padankan sejumlah metode penanganan *class imbalance* sehingga dihasilkan model dengan performa yang optimal. Pada tahap ini, model akan diatur *hyperparameter* `class_weight = 'balanced'`. Output dari bagian ini adalah dapat menemukan **threshold** terbaik, skema **upsampling** terbaik dan skema **downsampling** terbaik untuk model **Logistic Regression**.

```python
# membuat function yang mencari threshold terbaik
def best_hybrid(model, frac_list:list, multiplier_list:list[int], threshold_list:list[float], x_train, y_train, x_test, y_test):
    best_model = None
    best_frac = 0
    best_multiplier = 0
    best_threshold = 0
    best_f1_class_1 = 0
    
    for frac in frac_list:
        for multiplier in multiplier_list:
            for threshold in threshold_list:
                # hybrid upsampling & downsampling dataset
                x_train_ds, y_train_ds = downsampling(x_train=x_train, y_train=y_train, frac=frac)
                x_train_us, y_train_us = upsampling(x_train=x_train_ds, y_train=y_train_ds, multiplier=multiplier)
                
                # melatih model
                model.fit(x_train_us, y_train_us)
                prediction_prob = model.predict_proba(x_test)[:,1]
                
                # penyesuaian threshold
                predictions = prediction_prob>=threshold
                f1_class_1 = f1_score(y_test, predictions)
                if f1_class_1>best_f1_class_1:
                    best_model = model
                    best_f1_class_1 = f1_class_1
                    best_frac = frac
                    best_multiplier = multiplier
                    best_threshold = threshold

            
    print(f"Best F1 score class 1 : {best_f1_class_1:.2%}")
    print(f"Best fraction         : {best_frac}")
    print(f"Best multiplier       : {best_multiplier}")
    print(f"Best threshold        : {best_threshold}")
    
    return best_model
```

```python
# mencari setup hybrid terbaik
lr = LogisticRegression(solver='liblinear', class_weight='balanced', random_state=12345)
lr = best_hybrid(model=lr, 
            frac_list=np.arange(0.01, 0.5, 0.01),
            multiplier_list=[1,2,3,4],
            threshold_list=np.arange(0.1,1,0.1), 
            x_train=train_features, y_train=train_target, x_test=val_features, y_test=val_target) 
```

Kombinasi pengaturan penanganan *class imbalance* terbaik untuk model **Logistic Regression** memperoleh *F1 score* sebesar 53.85%. Untuk memperoleh model terbaik ini cukup menggunakan metode **downsampling** dengan fraksi sebesar 9%.

## <a id='toc6_3_'></a>[3) *Hyperparameter Tuning*](#toc0_)

### <a id='toc6_3_1_'></a>[**Logistic Regression**](#toc0_)

Pada model tuning menggunakan model **Logistic Regression**, terdapat sejumlah *hyperparameter* yang dapat diatur untuk meningkatkan kualitas model pada proyek ini, diantaranya `solver`, `max_iter` (jumlah iterasi maksimum), nilai `C` (*Regularization Parameter*), dan `Tol` (parameter konvergensi).

**Training Dataset** yang kita miliki berukuran kecil sehingga solver yang tepat untuk digunakan, yaitu `"liblinear"`.

Pada bagian sebelumnya, diperoleh informasi mengenai metode penanganan *class imbalance* pada **training dataset** dan berikut adalah metode yang harus dijalankan saat melatih model **Logistic Regression**.

* **Downsampling** dengan fraksi sebesar 9% dari data class 0.
* **Upsampling** dengan faktor pengali sebesar 1 dari data class 1.
* Pengaturan **threshold** menjadi senilai 0.5


```python
# Tuning hyperparameter
# pengaturan untuk penanganan class imbalance -- Logistic Regression
fraction = 0.09
multiplier = 1
threshold = 0.5

# mempersiapkan dataset yang telah di-resampling
## downsampling training dataset
x_train_lr, y_train_lr = downsampling(x_train=train_features, y_train=train_target, frac=fraction)
## upsampling training dataset
x_train_lr, y_train_lr = upsampling(x_train=x_train_lr, y_train=y_train_lr, multiplier=multiplier)

# melakukan model tuning pada Logistic Regression
max_iter_comb = [50, 100,150,200]
c_comb = [0.4, 0.45, 0.5, 0.55, 0.6]
tol_comb = [0.1, 0.01, 0.001, 0.0001]

# parameter tuning
best_lr = None
best_max_iter = 0
best_c_value = 0
best_tolerance = 0
best_f1_class_1 = 0

# iterasi model
for max_iter in max_iter_comb:
    for c_value in c_comb:
        for tolerance in tol_comb:
            lr = LogisticRegression(solver='liblinear', class_weight='balanced', random_state=12345,
                                    max_iter=max_iter, C=c_value, tol=tolerance,)
            lr.fit(x_train_lr, y_train_lr)
            prediction_prob = lr.predict_proba(val_features)[:,1]
            prediction = prediction_prob>=threshold
            f1_class_1 = f1_score(val_target, prediction)
            if f1_class_1>best_f1_class_1:
                best_lr = lr
                best_max_iter = max_iter
                best_c_value = c_value
                best_tolerance = tolerance
                best_f1_class_1 = f1_class_1

print(f"""
Best_max_iter          : {best_max_iter}
Best_c_value           : {best_c_value}
Best_tolerance         : {best_tolerance}
Best F1 score class 1  : {best_f1_class_1:.2%}
      """)
```

Berikut di bawah ini merupakan hasil evaluasi dari pengaturan *hyperparameter* terbaik pada model **Logistic Regression**.

```python
# menampilkan evaluasi model terbaik
model_evaluation(model=best_lr, x_test=val_features, y_test=val_target, x_train=train_features, y_train=train_target, threshold=0.5)
ConfMat_x_ROC(model=best_lr, x_test=val_features, y_test=val_target, threshold=0.5)
```

```python
# Melakukan pengujian model yang optimal pada testing dataset

# pengaturan untuk penanganan class imbalance -- Random Forest Classifier
fraction = 0.54
multiplier = 4
threshold = 0.55

# mempersiapkan dataset yang telah di-resampling
## downsampling training dataset
x_train_rf, y_train_rf = downsampling(x_train=train_features, y_train=train_target, frac=fraction)
## upsampling training dataset
x_train_rf, y_train_rf = upsampling(x_train=x_train_rf, y_train=y_train_rf, multiplier=multiplier)

# Instances Object RandomForestClassifier
final_model = RandomForestClassifier(min_samples_leaf=1 ,min_samples_split=2,
                                    max_depth=9, n_estimators=11,
                                    random_state=12345, class_weight="balanced")

# Melatih model
final_model.fit(x_train_rf, y_train_rf)

# prediksi training dataset
prediction_prob_train = final_model.predict_proba(train_features)[:,1]
prediction_train = prediction_prob_train>=threshold

# prediksi validation dataset
prediction_prob_val = final_model.predict_proba(val_features)[:,1]
prediction_val = prediction_prob_val>=threshold

# prediksi testing dataset
prediction_prob_test = final_model.predict_proba(test_features)[:,1]
prediction_test = prediction_prob_test>=threshold

# menghitung f1 score prediksi pada training, validation dan testing dataset
f1_class_1_train = f1_score(train_target, prediction_train)
f1_class_1_val = f1_score(val_target, prediction_val)
f1_class_1_test = f1_score(test_target, prediction_test)

# menghitung AUC-ROC prediksi pada training, validation dan testing dataset
auc_roc_train = roc_auc_score(train_target, prediction_train)
auc_roc_val = roc_auc_score(val_target, prediction_val)
auc_roc_test = roc_auc_score(test_target, prediction_test)

print(f"{' Evaluasi Metrik F1 Score ':-^100}")
print(f"F1 score model pada training dataset   : {f1_class_1_train:.2%}")
print(f"F1 score model pada validation dataset : {f1_class_1_val:.2%}")
print(f"F1 score model pada testing dataset    : {f1_class_1_test:.2%}")
print()
print(f"{' Evaluasi Metrik AUC-ROC ':-^100}")
print(f"AUC-ROC model pada training dataset   : {auc_roc_train:.2%}")
print(f"AUC-ROC model pada validation dataset : {auc_roc_val:.2%}")
print(f"AUC-ROC model pada testing dataset    : {auc_roc_test:.2%}")
```
